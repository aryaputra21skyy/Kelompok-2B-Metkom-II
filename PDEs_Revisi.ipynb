# **Persamaan Diferensial Parsial (PDEs)**

Pertama, beberapa notasi akan diperkenalkan sebagai langkah awal. Kita akan menggunakan notasi
$$
    \frac{\partial u}{\partial x} = u_x
$$
untuk menyatakan turunan parsial dari fungsi 𝑢 terhadap variabel 𝑥.

Turunan orde lebih tinggi maupun turunan campuran dinyatakan dengan cara yang serupa, yaitu
$$
    \frac{\partial^2 u}{\partial x^2} = u_{xx} \quad \quad \frac{\partial^2 u}{\partial y \partial x} = u_{xy}
$$

Perlu diperhatikan bahwa secara umum urutan pengambilan turunan dapat memengaruhi hasilnya.

Lalu, kapan turunan-turunan tersebut dapat dipertukarkan urutannya?

Pada bagian ini, kita menyoroti variabel-variabel yang umumnya berkaitan dengan ruang Kartesius. Namun, akan berguna juga untuk mempertimbangkan turunan dalam koordinat polar, seperti
$$
    \frac{\partial u}{\partial \theta} = u_\theta
$$
atau turunan terhadap waktu, yaitu
$$
    \frac{\partial u}{\partial t} = u_t.
$$    

Materi Persamaan Diferensial Parsial (PDE) bertujuan untuk memahami bagaimana fenomena geofisika dapat dimodelkan secara matematis melalui turunan parsial terhadap ruang dan waktu, seperti $( u_x )$, $( u_{xx} )$, dan $( u_t )$.

Dengan memahami notasi ini, mahasiswa dapat menggambarkan perubahan parameter fisik seperti suhu, gelombang, dan potensial dalam bumi. Selain itu, materi ini menjadi dasar dalam menyusun dan menyelesaikan model numerik menggunakan metode komputasi, seperti metode beda hingga.

Secara umum, pembelajaran PDE membantu dalam menghubungkan konsep matematika dengan simulasi fenomena geofisika secara lebih realistis.

## **1. Klasifikasi PDE**

Klasifikasi klasik persamaan diferensial parsial (PDE) didasarkan pada bentuk umum suatu PDE linear orde dua yang dapat dituliskan sebagai
$$
    a u_{xx} + 2b u_{xy} + c u_{yy} + d u_x + e u_y + f u = g.
$$

Perlu diperhatikan bahwa dalam kerangka umum klasifikasi ini, variabel 𝑥 atau 𝑦 juga dapat berupa 𝑡 (waktu) atau variabel lainnya.

Dengan melakukan transformasi Fourier secara formal terhadap persamaan 
$$
    a u_{xx} + 2b u_{xy} + c u_{yy} + d u_x + e u_y + f u = g.
$$
kita dapat mengubah persamaan umum tersebut menjadi
$$
    P(X,Y) = a X^2 + 2b X \cdot Y + c Y^2 + \cdots
$$
yang memiliki bentuk menyerupai polinom umum orde dua.

Istilah dalam klasifikasi PDE berasal dari fakta bahwa persamaan di atas merepresentasikan bentuk umum **irisan kerucut (*conic sections*)**, yang biasanya diklasifikasikan sebagai eliptik, parabolik, atau hiperbolik berdasarkan nilai diskriminannya,

$$
b^2 - ac.
$$

Selanjutnya, PDE diklasifikasikan berdasarkan tanda dari diskriminan tersebut, yaitu:

- jika $b^2 - ac < 0$, maka PDE bersifat eliptik,  
- jika $b^2 - ac = 0$, maka PDE bersifat parabolik, dan  
- jika $b^2 - ac > 0$, maka PDE bersifat hiperbolik.

Alasan mengapa klasifikasi ini penting adalah karena biasanya terdapat kelompok metode numerik tertentu yang bekerja dengan baik untuk satu jenis klasifikasi, tetapi tidak efektif untuk jenis lainnya.

Kita akan mempelajari lebih lanjut alasan dan mekanisme di balik hal tersebut ketika masing-masing jenis PDE diperkenalkan. Untuk saat ini, mari kita meninjau beberapa contoh dasar PDE dari setiap klasifikasi tersebut.

### **Contoh**

**Laplace's Equation:**  
Pada kasus berdimensi $n$, persamaan Laplace dituliskan sebagai

$$
\Delta u = \nabla^2 u = \sum^n_{i=1} u_{x_i x_i} = 0
$$

dan ketika $n = 2$ dengan $x$ dan $y$ sebagai dimensinya, diperoleh

$$
u_{xx} + u_{yy} = 0.
$$

Apa jenis PDE ini?

Dengan mengidentifikasi koefisien pada bentuk umum sebelumnya sebagai  
$a = 1$, $b = 0$, dan $c = 1$, maka diskriminannya adalah

$$
b^2 - ac = -1.
$$

Karena nilainya negatif, persamaan Laplace termasuk PDE bertipe **eliptik**.

```python
from IPython.display import display, Math

# Koefisien dari persamaan Laplace
a = 1
b = 0
c = 1

# Menghitung diskriminan
D = b**2 - a*c

# Menentukan jenis PDE
if D > 0:
    pde_type = "Hiperbolik"
elif D == 0:
    pde_type = "Parabolik"
else:
    pde_type = "Eliptik"

# Display persamaan dan hasil
display(Math(r"\text{Diketahui: } a = %d,\; b = %d,\; c = %d" % (a, b, c)))
display(Math(r"\text{Diskriminan: } D = b^2 - ac"))
display(Math(r"D = (%d)^2 - (%d)(%d) = %d" % (b, a, c, D)))
display(Math(r"\text{Jenis PDE: } \mathbf{%s}" % pde_type))
```

**Heat Equation:**  
Persamaan panas (*heat equation*) dalam satu dimensi ruang didefinisikan sebagai

$$
u_t = u_{xx}.
$$

Apa jenis PDE dari persamaan panas?

Persamaan panas merupakan PDE **parabolik**. Hal ini dapat dilihat dengan mengganti turunan terhadap $y$ pada bentuk umum sebelumnya dengan turunan terhadap waktu $t$, sehingga diperoleh koefisien

$$
a = -1, \quad b = 0, \quad c = 0,
$$

yang menghasilkan diskriminan bernilai nol.

Apa yang terjadi pada persamaan panas dalam beberapa dimensi ruang?

$$
u_t = u_{xx} + u_{yy}.
$$

Sayangnya, proses identifikasi klasifikasi menjadi tidak langsung dapat diterapkan di sini karena kita sekarang memiliki tiga variabel.

```python
from IPython.display import display, Math

# Heat Equation: u_t = u_xx

# Koefisien PDE
a = -1   # koefisien u_xx
b = 0    # koefisien u_xt
c = 0    # koefisien u_tt

# Menghitung diskriminan
D = b**2 - a*c

# Menentukan jenis PDE
if D > 0:
    pde_type = "Hiperbolik"
elif D == 0:
    pde_type = "Parabolik"
else:
    pde_type = "Eliptik"

# Display rapi dengan LaTeX
display(Math(r"\text{Persamaan Panas (Heat Equation): } u_t = u_{xx}"))
display(Math(r"\text{Koefisien: } a = %d,\; b = %d,\; c = %d" % (a, b, c)))
display(Math(r"\text{Diskriminan: } D = b^2 - ac"))
display(Math(r"D = (%d)^2 - (%d)(%d) = %d" % (b, a, c, D)))
display(Math(r"\text{Jenis PDE: } \mathbf{%s}" % pde_type))
```

**Wave Equation:**  
Persamaan gelombang (*wave equation*) dalam satu dimensi ruang didefinisikan sebagai

$$
u_{tt} - u_{xx} = 0.
$$

Melalui proses eliminasi, dapat diduga bahwa persamaan ini bersifat **hiperbolik**. Dengan kembali mengganti turunan terhadap $y$ pada bentuk umum dengan turunan terhadap waktu $t$, diperoleh koefisien

$$
a = -1, \quad b = 0, \quad c = 1,
$$

sehingga diskriminannya adalah

$$
b^2 - ac = -(-1)(1) = 1.
$$

Karena diskriminan bernilai positif, persamaan gelombang termasuk PDE bertipe **hiperbolik**.

```python
from IPython.display import display, Math

# Ringkasan klasifikasi tiga PDE terkenal
pde = {
    "Laplace Equation": (1, 0, 1),
    "Heat Equation": (-1, 0, 0),
    "Wave Equation": (-1, 0, 1)
}

for nama, (a, b, c) in pde.items():
    D = b**2 - a*c
    
    if D > 0:
        jenis = "Hiperbolik"
    elif D == 0:
        jenis = "Parabolik"
    else:
        jenis = "Eliptik"
    
    # Display 
    display(Math(r"\textbf{%s}" % nama))
    display(Math(r"\text{Koefisien: } a = %d,\; b = %d,\; c = %d" % (a, b, c)))
    display(Math(r"\text{Diskriminan: } D = b^2 - ac"))
    display(Math(r"D = (%d)^2 - (%d)(%d) = %d" % (b, a, c, D)))
    display(Math(r"\text{Jenis PDE: } \mathbf{%s}" % jenis))
    display(Math(r"\text{------------------------------}"))
```

**Persamaan Navier–Stokes:**  
Persamaan Navier–Stokes memodelkan aliran fluida tak-mampat (*incompressible fluid flow*)

$$
\begin{aligned}
u_t + (u^2)_x + (uv)_y &= - \frac{1}{\rho} p_x + \nu (u_{xx} + v_{yy}) \\\\
v_t + (uv)_x + (v^2)_y &= - \frac{1}{\rho} p_y + \nu (u_{xx} + v_{yy})
\end{aligned}
$$

Meskipun kita mengabaikan fakta bahwa persamaan-persamaan ini tidak dapat diklasifikasikan secara langsung menggunakan sistem penamaan sebelumnya, persamaan tersebut mengandung bagian yang menyerupai komponen **hiperbolik** (pada ruas kiri) serta komponen **eliptik** (pada ruas kanan), atau bahkan **parabolik** jika mempertimbangkan turunan terhadap waktu.

Karena alasan ini, persamaan Navier–Stokes sering dianggap sebagai PDE bertipe **parabolik**. Namun, pendekatan penyelesaiannya dalam praktik sering kali menggunakan teknik dari masing-masing klasifikasi PDE.

Salah satu cara yang lebih umum dan bermanfaat untuk memahami klasifikasi PDE adalah dengan meninjau bagaimana informasi merambat di dalam domain.

- **PDE Hiperbolik:** informasi merambat dengan kecepatan hingga (kecepatan grup bernilai riil dan terbatas).

- **PDE Parabolik:** menggambarkan pergerakan informasi yang berlangsung dengan kecepatan tak hingga (misalnya pada proses difusi dalam ruang hampa).

- **PDE Eliptik:** merupakan kasus batas (*limiting case*) di antara persamaan hiperbolik dan parabolik dalam hal perambatan informasi.

## **2. Analisis Fourier**

Analisis Fourier merupakan salah satu alat dasar dalam menganalisis dan menyelesaikan PDE linear. Pada domain hingga (*finite domains*), kita biasanya menggunakan deret Fourier untuk merepresentasikan solusi. Penggunaan deret Fourier setara dengan memandang fungsi eigen (*eigenfunctions*) dari operator diferensial sebagai basis dari solusi.

Interpretasi PDE melalui sudut pandang aljabar linear ini sangat kuat, dan konsep ini akan terus muncul berulang kali dalam pembahasan selanjutnya.

### **Deret Fourier (Fourier Series)**

Deret Fourier suatu fungsi memiliki beberapa bentuk berbeda, bergantung pada jenis fungsi yang ditinjau. Ingat bahwa $\sin x$ merupakan fungsi ganjil (*odd function*) dan $\cos x$ merupakan fungsi genap (*even function*).

Pada interval hingga:

- untuk fungsi ganjil, kita menggunakan deret Fourier sinus,
- untuk fungsi genap, kita menggunakan deret Fourier kosinus.

Jika suatu fungsi bukan ganjil maupun genap, maka digunakan deret Fourier lengkap yang biasanya dituliskan sebagai

$$
f(x) = a_0 + \sum^\infty_{n=1} a_n \cos \left( \frac{n \pi x}{L} \right)
+ \sum^\infty_{n=1} b_n \sin \left( \frac{n \pi x}{L} \right),
$$

dengan koefisien-koefisien

$$
\begin{aligned}
a_0 &= \frac{1}{L} \int^L_0 f(x)\, dx, \\\\
a_n &= \frac{2}{L} \int^L_0 f(x) \cos \frac{n \pi x}{L}\, dx, \\\\
b_n &= \frac{2}{L} \int^L_0 f(x) \sin \frac{n \pi x}{L}\, dx.
\end{aligned}
$$

Koefisien-koefisien ini merupakan proyeksi fungsi $f(x)$ terhadap basis Fourier berupa fungsi $\cos$ dan $\sin$, yang membentuk basis ortogonal.

Bagaimana konsep ini berkaitan dengan aljabar linear dan ruang vektor berdimensi hingga (*finite-dimensional vector spaces*)?

### **Transformasi Fourier (Fourier Transform)**

Transformasi Fourier dari suatu fungsi $v(x)$ didefinisikan sebagai

$$
\widehat{v}(\xi) = \frac{1}{\sqrt{2 \pi}} \int^\infty_{-\infty} v(x)\, e^{-i \xi x}\, dx.
$$

Agar transformasi ini ada (terdefinisi), fungsi harus memenuhi $v \in L^2$, yaitu

$$
||v||_2 = \left ( \int^\infty_{-\infty} |v(x)|^2 dx \right )^{1/2} < \infty.
$$

Fungsi $\widehat{v}(\xi)$ juga berada dalam ruang $L^2$ karena berdasarkan identitas Parseval diperoleh

$$
||v(x)||_2 = ||\widehat{v}(\xi)||_2.
$$

Kita juga dapat menggunakan transformasi invers untuk memperoleh kembali fungsi asal:

$$
v(x) = \frac{1}{\sqrt{2 \pi}} \int^\infty_{-\infty} \widehat{v}(\xi)\, e^{i \xi x}\, d\xi.
$$

Salah satu sifat penting Transformasi Fourier yang telah digunakan sebelumnya adalah bahwa transformasi Fourier mengubah turunan spasial menjadi hubungan aljabar:

$$
\widehat{v_x} = i \xi \widehat{v}.
$$

Hal ini terjadi karena fungsi

$$
e^{i \xi x}
$$

merupakan fungsi eigen (*eigenfunctions*) dari operator

$$
\frac{\partial}{\partial x}.
$$

Konsep ini juga berlaku pada kasus diskret, di mana analog diskret dari fungsi eigen akan menjadi vektor eigen (*eigenvectors*) dari operator beda hingga (*finite difference operators*).

**Latihan:**  
Cobalah mengambil transformasi Fourier dari persamaan adveksi

$$
u_t + c u_x = 0.
$$

Karena turunan terhadap waktu tidak dipengaruhi oleh transformasi Fourier, diperoleh

$$
\widehat{u}_t = -i \xi c \widehat{u},
$$

yang merupakan sistem ODE tak hingga.

Sekarang, selesaikan sistem tersebut untuk mendapatkan solusi persamaan adveksi.

Kita mengetahui bahwa solusi ODE berbentuk

$$
\widehat{u}(\xi, t) = \widehat{u}(\xi, 0)\, e^{-i \xi c t}.
$$

Dengan menggunakan transformasi invers, diperoleh

$$
\begin{aligned}
u(x,t) &= \frac{1}{\sqrt{2 \pi}} \int^\infty_{-\infty}
\widehat{u}(\xi, 0)\, e^{-i \xi c t} e^{i \xi x} d\xi, \\\\
&= \frac{1}{\sqrt{2 \pi}} \int^\infty_{-\infty}
\widehat{u}(\xi, 0)\, e^{i \xi (x - c t)} d\xi, \\\\
&= u(x - c t, 0).
\end{aligned}
$$

## **3. Persamaan Eliptik (Elliptic Equations)**

Persamaan eliptik paling sering muncul sebagai masalah nilai batas (*boundary value problems*) pada suatu domain. Sebelumnya telah disebutkan persamaan Laplace, yang merupakan kasus khusus dari persamaan eliptik klasik lainnya, yaitu masalah Poisson

$$
\nabla^2 u = f.
$$

Jenis permasalahan ini sering muncul ketika kita ingin mencari keadaan tunak (*steady-state*) dari suatu masalah yang bergantung waktu, seperti persamaan panas (*heat equation*). Persamaan ini juga dapat muncul pada masalah bergantung waktu ketika terdapat dinamika yang jauh lebih cepat dibandingkan skala waktu proses lainnya. Salah satu contohnya adalah kondisi ketakmampatan (*incompressibility condition*) pada persamaan Navier–Stokes, yang menghasilkan masalah Poisson untuk tekanan.

Perilaku utama dari masalah nilai batas eliptik adalah bahwa solusi di bagian dalam domain $\Omega$ sepenuhnya ditentukan oleh kondisi batas dan fungsi sumber $f$. Hal ini sangat penting dalam metode numerik, karena penyelesaiannya sering melibatkan sistem persamaan linear berukuran besar dan jarang (*large sparse linear systems*), di mana setiap titik saling terhubung dengan titik lainnya.

Secara umum, suatu operator linear $L$ berbentuk

$$
L = \sum^n_{i,j=1} A_{ij} \frac{\partial^2}{\partial x_i \partial x_j}
+ \sum^n_{i=1} B_i \frac{\partial}{\partial x_i}
+ C
$$

dapat disebut **eliptik** apabila matriks $A$ bersifat definit positif atau definit negatif, yaitu ketika

$$
v^T A v
$$

memiliki tanda yang sama untuk setiap $v \neq 0$ dengan $v \in \mathbb{R}^n$.

Karena kita memiliki kondisi batas tak-homogen pada setiap sisi persegi panjang, kita tidak dapat langsung menggunakan metode pemisahan variabel (*separation of variables*). Oleh karena itu, masalah diuraikan menjadi empat sub-bagian sehingga

$$
u(x,y) = u_1(x,y) + u_2(x,y) + u_3(x,y) + u_4(x,y),
$$

di mana setiap $u_i(x,y)$ mempertahankan satu sisi tak-homogen dari persegi panjang, sementara sisi lainnya dibuat bernilai nol.

Sebagai contoh, untuk $u_1(x,y)$ diperoleh

$$
\begin{aligned}
\nabla^2 u_1 = 0 \quad & \quad \Omega = (0, L) \times (0, H) \\\\
u_1(x,0) &= f_1(x) \\\\
u_1(x,H) &= 0 \\\\
u_1(0,y) &= 0 \\\\
u_1(L,y) &= 0.
\end{aligned}
$$

Dengan menggunakan metode pemisahan variabel, diasumsikan solusi berbentuk

$$
u_4(x,y) = h(x)\,\phi(y),
$$

yang ketika disubstitusikan ke dalam PDE menghasilkan

$$
\begin{aligned}
&\phi \frac{\mathrm{d}^2 h}{\mathrm{d} x^2}
+ h \frac{\mathrm{d}^2 \phi}{\mathrm{d} y^2} = 0 \\\\
&\frac{1}{h} \frac{\mathrm{d}^2 h}{\mathrm{d} x^2}
= - \frac{1}{\phi} \frac{\mathrm{d}^2 \phi}{\mathrm{d} y^2}
= -\lambda \Rightarrow \\\\
&h''(x) + \lambda h(x) = 0,
\quad\quad
\phi''(y) - \lambda \phi(y) = 0.
\end{aligned}
$$

Kondisi batas untuk kedua fungsi tersebut adalah

$$
h(0) = h(L) = 0,
\quad
\phi(0) = f_1(x),
\quad
\phi(H) = 0.
$$

Pertama, dengan menyelesaikan masalah nilai batas homogen yang lebih sederhana untuk $h(x)$, diperoleh

$$
h(x) = A \cos (\sqrt{\lambda}\, x) + B \sin(\sqrt{\lambda}\, x).
$$

Dengan menerapkan kondisi batas diperoleh

$$
\begin{aligned}
&h(0) = 0 \Rightarrow A = 0, \\\\
&h(L) = 0 \Rightarrow B \sin(\sqrt{\lambda}\, L) = 0.
\end{aligned}
$$

Untuk memenuhi kondisi batas terakhir terdapat dua kemungkinan:  
kita dapat memilih $B = 0$ yang menghasilkan solusi trivial, atau

$$
\sqrt{\lambda}\, L = n \pi.
$$

Dengan menggunakan kondisi kedua diperoleh nilai eigen

$$
\lambda_n = \left(\frac{n \pi}{L}\right)^2,
\quad n = 1,2,\ldots
$$

sehingga solusi yang diperoleh adalah

$$
h_n(x) = B_n \sin\left(\frac{n \pi x}{L}\right).
$$

Untuk persamaan $\phi$, diperoleh solusi berbentuk

$$
\phi(y) = A e^{\sqrt{\lambda} y} + B e^{-\sqrt{\lambda} y}
$$

atau secara ekuivalen dapat dituliskan sebagai

$$
\phi(y) = A \cosh(\sqrt{\lambda}\, y) + B \sinh(\sqrt{\lambda}\, y).
$$

Untuk memenuhi kondisi batas homogen, solusi tersebut perlu sedikit dimodifikasi dengan melakukan pergeseran fungsi sehingga menjadi

$$
\phi(y) = A \cosh \left(\sqrt{\lambda} (y - H)\right)
+ B \sinh \left(\sqrt{\lambda} (y - H)\right).
$$

Dengan bentuk ini, kondisi batas homogen dapat dipenuhi melalui

$$
\begin{aligned}
\phi(H) = 0
&= A \cosh \left(\sqrt{\lambda}(H-H)\right)
+ B \sinh \left(\sqrt{\lambda}(H-H)\right) \\\\
&= A \cosh(0) + B \sinh(0) \\\\
&= A,
\end{aligned}
$$

sehingga diperoleh $A = 0$ dan solusi menjadi

$$
\phi(y) = B \sinh \left(\frac{n \pi}{L}(y - H)\right).
$$

Dengan menggabungkan kedua solusi, diperoleh

$$
u_{1,n}(x,y)
= B_n \sin\left(\frac{n \pi x}{L}\right)
\sinh \left(\frac{n \pi}{L}(y - H)\right).
$$

Melalui prinsip superposisi, solusi umum menjadi

$$
u_1(x,y)
= \sum^\infty_{n=1}
B_n \sin\left(\frac{n \pi x}{L}\right)
\sinh \left(\frac{n \pi}{L}(y - H)\right).
$$

Selanjutnya, untuk memenuhi kondisi batas terakhir dan menentukan koefisien $B_n$, digunakan

$$
u_1(x,0) = f_1(x)
= \sum^\infty_{n=1}
- B_n \sin\left(\frac{n \pi x}{L}\right)
\sinh \left(\frac{n \pi H}{L}\right).
$$

Dengan memanfaatkan sifat ortogonalitas fungsi sinus diperoleh

$$
\begin{aligned}
\int^L_0 f_1(x)
\sin\left(\frac{m \pi x}{L}\right)\, dx
&=
\sum^\infty_{n=1}
- B_n \sinh\left(\frac{n \pi H}{L}\right)
\int^L_0
\sin\left(\frac{n \pi x}{L}\right)
\sin\left(\frac{m \pi x}{L}\right)\, dx \\\\
\int^L_0 f_1(x)
\sin\left(\frac{m \pi x}{L}\right)\, dx
&=
- B_m \frac{L}{2}
\sinh\left(\frac{m \pi H}{L}\right),
\end{aligned}
$$

sehingga koefisien diperoleh sebagai

$$
B_m =
\frac{-2}{
L \sinh \left(\frac{m \pi H}{L}\right)
}
\int^L_0
f_1(x)
\sin\left(\frac{m \pi x}{L}\right)\, dx.
$$

Dengan demikian, solusi untuk komponen $u_1$ dari masalah umum di atas telah diperoleh.

### **Contoh Penerapan Persamaan Eliptik**

#### **Distribusi Suhu Keadaan Tunak (Steady-State Heat Conduction)**

Misalkan suatu pelat tipis dua dimensi pada domain $\Omega$ memiliki distribusi suhu $u(x,y)$ yang tidak berubah terhadap waktu. Kondisi ini disebut sebagai keadaan tunak (*steady-state*), sehingga model matematisnya diberikan oleh persamaan Laplace:

$$
\nabla^2 u = 0, \quad (x,y) \in \Omega
$$

atau secara eksplisit:

$$
\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 0
$$

Pada kasus ini, tidak terdapat sumber atau penyerap panas di dalam domain, sehingga distribusi suhu sepenuhnya ditentukan oleh kondisi batas yang diberikan.

Sebagai contoh, tinjau domain berbentuk persegi dengan $0 < x < L$ dan $0 < y < L$, serta kondisi batas:

$$
u(x,0) = 0, \quad u(x,L) = 100
$$
$$
u(0,y) = 0, \quad u(L,y) = 0
$$

Secara matematis, solusi $u(x,y)$ yang diperoleh akan merepresentasikan distribusi suhu di seluruh pelat. Nilai suhu di setiap titik dalam domain ditentukan oleh interaksi dengan titik-titik di sekitarnya, sehingga solusi yang dihasilkan bersifat halus dan kontinu.

Persamaan Laplace memiliki sifat penting, yaitu nilai solusi pada suatu titik sama dengan rata-rata nilai di sekitarnya. Hal ini menunjukkan bahwa sistem berada dalam kondisi keseimbangan, di mana tidak terjadi perubahan energi terhadap waktu.

Jika di dalam domain terdapat sumber panas, maka modelnya berubah menjadi persamaan Poisson:

$$
\nabla^2 u = f(x,y)
$$

Dalam hal ini, fungsi $f(x,y)$ menyatakan distribusi sumber atau penyerap panas. Keberadaan $f$ menyebabkan distribusi suhu tidak lagi murni ditentukan oleh batas, tetapi juga dipengaruhi oleh kondisi internal domain.

Dengan demikian, persamaan eliptik secara umum digunakan untuk memodelkan fenomena yang bersifat **keseimbangan ruang**, di mana solusi di dalam domain sangat dipengaruhi oleh kondisi batas dan, jika ada, oleh sumber di dalam domain tersebut.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')

plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']
plt.rcParams['axes.edgecolor'] = '#1f3b73'
plt.rcParams['axes.labelcolor'] = '#cfe2ff'
plt.rcParams['xtick.color'] = '#cfe2ff'
plt.rcParams['ytick.color'] = '#cfe2ff'
plt.rcParams['text.color'] = '#e6f0ff'

nx, ny = 50, 50
Lx, Ly = 1.0, 1.0

dx = Lx / (nx - 1)
dy = Ly / (ny - 1)

u = np.zeros((ny, nx))

# Kondisi batas
u[:, 0] = 0        # kiri
u[:, -1] = 0       # kanan
u[0, :] = 0        # bawah
u[-1, :] = 100     # atas

max_iter = 2000

for k in range(max_iter):
    u_old = u.copy()
    
    for i in range(1, ny-1):
        for j in range(1, nx-1):
            u[i, j] = 0.25 * (
                u[i+1, j] + u[i-1, j] +
                u[i, j+1] + u[i, j-1]
            )
    
    error = np.max(np.abs(u - u_old))
    if error < 1e-5:
        print(f"Konvergen pada iterasi ke-{k}")
        break

plt.figure(figsize=(7, 6))

im = plt.imshow(
    u,
    origin='lower',
    extent=[0, Lx, 0, Ly],
    cmap='Blues'   # nuansa biru
)

cbar = plt.colorbar(im)
cbar.set_label('Suhu', fontsize=11)

plt.title('Distribusi Suhu Steady-State (Laplace 2D)', fontsize=14, fontweight='bold')
plt.xlabel('Posisi x', fontsize=12)
plt.ylabel('Posisi y', fontsize=12)

# Grid halus
plt.grid(alpha=0.1, linestyle='--')

# Memperhalus border
for spine in plt.gca().spines.values():
    spine.set_linewidth(1.2)

plt.tight_layout()
plt.show()
```

## **4. Persamaan Parabolik**
Persamaan parabolik mirip dengan persamaan eliptik namun sering juga menyertakan ketergantungan waktu (ini tidak selalu terjadi tetapi akan
berlaku untuk pertimbangan kita). Misalnya kita dapat mengambil operator linear eliptik seperti yang didefinisikan di atas dan membuat persamaan
parabolik dengan menulis
$$
    u_t - Lu = f.
$$
Contoh klasik persamaan parabolik adalah persamaan panas (atau difusi)
yang didefinisikan sebagai
$$
    u_t = \nabla^2 u.
$$
PDP ini membutuhkan kondisi awal $u(x, 0)$ serta kondisi batas pada $\Omega$.

### **Menyelesaikan Persamaan Parabolik**   
Seperti pada masalah eliptik kita kembali menggunakan deret Fourier untuk
menyelesaikan persamaan parabolik. Mari lagi mempertimbangkan masalah model
kita, persamaan panas, dalam satu dimensi ruang:
$$\begin{aligned}
    &u_t = \kappa u_{xx}, \quad \Omega = (0, T) \times (0, L), \\
    &u(x, 0) = f(x), \\
    &u(0, t) = 0, \quad \text{dan} \\
    &u(L, t) = 0.
\end{aligned}$$
Dengan pemisahan variabel kita mengasumsikan $u(x,t) = G(t) \phi(x)$ dan
dapat menuliskan dua ODE
$$\begin{aligned}
    &\frac{\text{d} G}{\text{d} t} = -\kappa \lambda G, \\
    &\phi''(x) + \lambda \phi(x) = 0.
\end{aligned}$$
ODE yang bergantung waktu memiliki solusi berbentuk
$$
    G(t) = C e^{-\lambda \kappa t}.
$$
BVP untuk $\phi$ memiliki solusi berbentuk
$$
    \phi(x) = A \cos \left( \sqrt{\lambda} x \right) + B \sin \left( \sqrt{\lambda} x \right).
$$
Menggunakan kondisi batas kita temukan bahwa $A = 0$ dan untuk solusi
non-trivial bahwa
$$
    \lambda = \left( \frac{n \pi}{L} \right )^2,
$$
sehingga menghasilkan solusi berbentuk
$$
    u_n(x,t) = B_n e^{-\left(\frac{n \pi}{L}\right)^2 \kappa t} \sin \left( \frac{n \pi x}{L} \right),
$$
dan dengan superposisi kita punya
$$
    u(x,t) = \sum^\infty_{i=1} B_n e^{-\left(\frac{n \pi}{L}\right)^2 \kappa t} \sin \left( \frac{n \pi x}{L} \right).
$$
Dengan mengevaluasi kondisi awal kita memperoleh
$$
    u(x,0) = f(x) = \sum^\infty_{i=1} B_n \sin \left( \frac{n \pi x}{L} \right),
$$
yang memberi $B_n$ dengan kembali menggunakan ortogonalitas fungsi $\sin$
$$\begin{aligned}
    \int^L_0 f(x) \sin \left( \frac{m \pi x}{L} \right) dx &= \sum^\infty_{i=1} B_n \int^L_0\sin \left( \frac{n \pi x}{L} \right) \sin \left( \frac{m \pi x}{L} \right) dx, \\
    \int^L_0 f(x) \sin \left( \frac{m \pi x}{L} \right) dx &= B_m \frac{L}{2}, \\
    B_m &= \frac{2}{L} \int^L_0 f(x) \sin \left( \frac{m \pi x}{L} \right) dx,
\end{aligned}$$
sehingga menyelesaikan PDP.

### **Contoh Penerapan Persamaan Parabolik**
### **Perpindahan Panas pada Batang 1D**

Misalkan suatu batang homogen dengan panjang $L$ mengalami proses perpindahan panas. Distribusi suhu pada batang dinyatakan oleh fungsi $u(x,t)$, yang memenuhi persamaan panas:

$$
u_t = \kappa u_{xx}, \quad 0 < x < L,\; t > 0
$$

dengan kondisi awal dan batas:
$$
u(x,0) = f(x), \quad 0 \le x \le L
$$
$$
u(0,t) = 0, \quad u(L,t) = 0
$$

Persamaan ini menyatakan bahwa laju perubahan suhu terhadap waktu ($u_t$) sebanding dengan kelengkungan distribusi suhu terhadap ruang ($u_{xx}$). Secara fisik, hal ini menggambarkan bahwa panas mengalir dari daerah bersuhu tinggi ke rendah.

**Solusi dengan Pemisahan Variabel**

Untuk menyelesaikan persamaan, digunakan metode pemisahan variabel dengan asumsi:

$$
u(x,t) = X(x)T(t)
$$

Substitusi ke persamaan menghasilkan:

$$
X(x)T'(t) = \kappa X''(x)T(t)
$$

Dengan memisahkan variabel diperoleh:

$$
\frac{T'}{\kappa T} = \frac{X''}{X} = -\lambda
$$

di mana $\lambda$ adalah konstanta pemisahan. Hal ini menghasilkan dua persamaan diferensial biasa (ODE):

$$
T'(t) + \kappa \lambda T(t) = 0
$$

$$
X''(x) + \lambda X(x) = 0
$$

**Masalah Nilai Batas (BVP)**

Fungsi ruang $X(x)$ harus memenuhi syarat batas:

$$
X(0)=0,\quad X(L)=0
$$

Masalah ini merupakan masalah nilai batas yang menghasilkan solusi non-trivial hanya jika:

$$
\lambda_n = \left(\frac{n\pi}{L}\right)^2,\quad n=1,2,3,\dots
$$

dengan fungsi eigen:

$$
X_n(x) = \sin\left(\frac{n\pi x}{L}\right)
$$

Fungsi-fungsi ini membentuk basis ortogonal yang digunakan untuk menyusun solusi umum.

**Solusi Bagian Waktu**

Persamaan waktu:

$$
T'(t) + \kappa \lambda_n T(t) = 0
$$

memiliki solusi:

$$
T_n(t) = e^{-\kappa \lambda_n t}
$$

Solusi ini menunjukkan bahwa setiap mode akan mengalami peluruhan eksponensial terhadap waktu.

**Solusi Umum**

Dengan menggabungkan solusi ruang dan waktu, diperoleh solusi umum dalam bentuk deret Fourier:

$$
u(x,t) = \sum_{n=1}^{\infty} B_n e^{-\kappa \left(\frac{n\pi}{L}\right)^2 t} \sin\left(\frac{n\pi x}{L}\right)
$$

Koefisien $B_n$ ditentukan dari kondisi awal:

$$
u(x,0) = f(x)
$$

sehingga:

$$
B_n = \frac{2}{L}\int_0^L f(x)\sin\left(\frac{n\pi x}{L}\right)\,dx
$$

Koefisien ini merepresentasikan kontribusi masing-masing mode gelombang terhadap distribusi suhu awal.


```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('dark_background')

plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']
plt.rcParams['axes.edgecolor'] = '#1f3b73'
plt.rcParams['axes.labelcolor'] = '#cfe2ff'
plt.rcParams['xtick.color'] = '#cfe2ff'
plt.rcParams['ytick.color'] = '#cfe2ff'
plt.rcParams['text.color'] = '#e6f0ff'

L = 1.0
T = 0.2
nx = 50
nt = 200
kappa = 1.0

dx = L / (nx - 1)
dt = T / nt

r = kappa * dt / dx**2
print("Nilai r =", r)

x = np.linspace(0, L, nx)
u = np.sin(np.pi * x)

hasil = []

for n in range(nt):
    u_new = u.copy()
    
    for i in range(1, nx-1):
        u_new[i] = u[i] + r * (u[i+1] - 2*u[i] + u[i-1])
    
    # Kondisi batas
    u_new[0] = 0
    u_new[-1] = 0
    
    u = u_new.copy()
    
    if n % 40 == 0:
        hasil.append(u.copy())

plt.figure(figsize=(8, 5))

colors = ['#4dabf7', '#339af0', '#228be6', '#1c7ed6', '#1971c2']  # gradasi biru

for i, h in enumerate(hasil):
    plt.plot(
        x, h,
        color=colors[i % len(colors)],
        linewidth=2.5,
        label=f"t = {i*(T/len(hasil)):.2f}"
    )

plt.xlabel("Posisi x", fontsize=12)
plt.ylabel("Suhu u(x,t)", fontsize=12)
plt.title("Evolusi Suhu Batang (Persamaan Panas)", fontsize=14, fontweight='bold')

plt.legend(frameon=False)
plt.grid(alpha=0.2, linestyle='--')

# Tambahan estetika
for spine in plt.gca().spines.values():
    spine.set_linewidth(1.2)

plt.tight_layout()
plt.show()
```

## **5. Persamaan Hiperbolik**   
Persamaan diferensial parsial (PDP) tipe hiperbolik banyak digunakan untuk memodelkan fenomena fisika yang melibatkan perambatan gelombang, serta berkaitan dengan hukum kekekalan dan hukum keseimbangan.
Contohnya:
- Kekekalan momentum (Hukum II Newton)
- Kekekalan energi

Pada prinsipnya, hukum-hukum tersebut menggambarkan bagaimana suatu besaran fisik berubah terhadap waktu dan ruang secara simultan. Contoh paling umum dari persamaan hiperbolik adalah persamaan gelombang:

$$
u_{tt} - c^2 u_{xx} = 0
$$

**Keterangan:**
- $u(x,t)$ : fungsi gelombang
- $u_{tt}$ : turunan kedua terhadap waktu (DTT - Turunan Terhadap Waktu)
- $u_{xx}$ : turunan kedua terhadap ruang
- $c$ : kecepatan rambat gelombang

**Makna DTT (Turunan Terhadap Waktu)**

DTT merupakan turunan terhadap variabel waktu $t$, yang memiliki makna fisik penting:

- $u_t = \frac{\partial u}{\partial t}$ → menyatakan kecepatan perubahan gelombang
- $u_{tt} = \frac{\partial^2 u}{\partial t^2}$ → menyatakan percepatan gelombang

Dalam konteks fisika, $u_{tt}$ berkaitan langsung dengan Hukum II Newton:

$$
F = m a
$$

di mana percepatan $a$ direpresentasikan oleh turunan kedua terhadap waktu. Oleh karena itu, persamaan gelombang dapat dipandang sebagai bentuk matematis dari hukum dinamika tersebut.

**Transformasi ke Sistem Orde Satu**

Dalam analisis numerik, persamaan hiperbolik orde dua seperti persamaan gelombang sering diubah menjadi sistem persamaan orde satu untuk mempermudah penyelesaian.

Bentuk umumnya adalah:

$$
u_t + A u_x = 0
$$

**Keterangan:**
- $u$ : vektor variabel (misalnya terdiri dari beberapa fungsi)
- $A$ : matriks koefisien

**Syarat Persamaan Hiperbolik**

Suatu sistem PDP dikatakan hiperbolik jika memenuhi:
- Nilai eigen $\lambda$ dari matriks $A$ adalah **real
- Matriks $A$ dapat **didiagonalisasi

### **Makna Fisik Sifat Hiperbolik**

Sifat ini menunjukkan bahwa:
- Gelombang merambat dengan **kecepatan nyata
- Informasi menyebar sepanjang **karakteristik
- Sistem memiliki perilaku stabil secara fisik

### **Menyelesaikan Persamaan Hiperbolik**   
Pendekatan paling umum untuk menyelesaikan PDP hiperbolik adalah
menemukan karakteristik dari masalah yang diberikan dan menggambar pada
bidang ($x$,$t$) tempat-tempat di mana sesuatu yang menarik mungkin terjadi.
Jika persamaan non-linear ini dapat menghasilkan kejutan (diskontinuitas)
meskipun kondisi awalnya halus. Ini akan menjadi masalah secara numerik
seperti yang akan kita lihat nanti dalam mata kuliah ini.

Pendekatan dasar metode karakteristik menggunakan asumsi bahwa solusi kita
berbentuk
$$
    u(x,t) = u(x(t), t),
$$
artinya solusi dapat dideskripsikan oleh satu variabel karakteristik, dalam
hal ini $t$. Misalkan kita ingin menemukan solusi dari
$$
    u_t + c(x,t,u) u_x = 0.
$$
Jika kita mengambil turunan penuh (total) dari solusi baru ini kita memperoleh
$$
    \frac{\text{d} u}{\text{d} t} = \frac{\partial u}{\partial t} + \frac{\text{d} x}{\text{d} t} \frac{\partial u}{\partial x},
$$
yang setelah mentransformasikan persamaan asli untuk menggunakan $\xi$,
mengarah pada kesimpulan bahwa PDP asli dapat ditulis ulang menjadi dua ODE
$$\begin{aligned}
    &\frac{\text{d} u}{\text{d} t} = 0, \quad \text{dan}\\
    &\frac{\text{d} x}{\text{d} t} = c(x,t,u).
\end{aligned}$$
Sekarang kita tahu bahwa $u(x(t),t)$ konstan sepanjang kurva $x(t)$ dan
sebenarnya
$$
    u(x,t) = u(x(t), 0).
$$

Solusi PDP kemudian bergantung pada penyelesaian kurva karakteristik yang
diberikan oleh $x(t)$.

Sebagai contoh sederhana mari selesaikan persamaan adveksi linear pada
domain tak hingga
$$\begin{aligned}
    &u_t + c u_x = 0, \quad \Omega = (-\infty, \infty) \times (0, T) \\
    &u(x,0) = f(x), \quad \text{dan} \\
    &u_x(\pm \infty, t) \rightarrow 0.
\end{aligned}$$
Persamaan karakteristik adalah
$$
    \frac{\text{d} x}{\text{d} t} = c,
$$
yang solusinya $x(t) = c t + x_0$. Kita kemudian dapat menggambar
karakteristik sebagai


di mana persamaan garisnya adalah $x(t) = c t + x_0$.

Untuk persamaan nonlinear, yakni ketika kecepatan dapat bergantung pada solusi
$c(u)$, kondisi awal akan menentukan kemiringan karakteristik sehingga
mengarah pada dua kasus. Mari pertimbangkan persamaan Burger
$$
    u_t + u u_x = 0,
$$
dengan pengaturan yang sama seperti PDP linear di atas. Coba lihat apakah
Anda dapat menemukan kurva karakteristik untuk kondisi awal
$$
    u(x,0) = \left \{ \begin{aligned}
        4, \;\; & x<0, \\
        3, \;\; & x>0.
    \end{aligned} \right.
$$
Apa karakteristik untuk masalah ini? Apa masalah yang muncul dalam kasus ini?
Kurva karakteristiknya adalah
$$
    x(t) = \left \{ \begin{aligned}
        4 t + x_0, \;\; & x<0, \\
        3 t + x_0, \;\; & x>0,
    \end{aligned} \right.
$$
yang menyebabkan tumbukan kurva karakteristik dan karena itu terbentuk
kejutan (shock).

Kita dapat menemukan persamaan untuk lokasi kejut menggunakan kondisi
Rankine--Hugoniot
$$
    \frac{\text{d} x_s}{\text{d} t} = \frac{f(x_+, t) - f(x_-, t)}{u_+ - u_-},
$$
di mana $x_s(t)$ adalah lokasi kejut, $f$ adalah fungsi fluks yang
didefinisikan dari $u_t + f(u)_x = 0$, dan notasi $u_\pm$ menyatakan nilai
di kiri dan kanan kejut bergantung tanda.

Bagaimana jika kita memiliki kondisi awal
$$
    u(x,0) = \left \{ \begin{aligned}
        3, \;\; & x<0, \\
        4, \;\; & x>0,
    \end{aligned} \right.
$$
sebagai gantinya? Jenis solusi apa yang kita harapkan sekarang?
Kurva karakteristik sekarang
$$
    x(t) = \left \{ \begin{aligned}
        3 t + x_0, \;\; & x<0, \\
        4 t + x_0, \;\; & x>0,
    \end{aligned} \right.
$$
menyisakan bagian bidang $(x,t)$ tanpa karakteristik. Sebagai gantinya kita
mengasumsikan solusi yang berubah halus, disebut rarefaksi, dan mengisi
solusi yang hilang.

Dalam kasus ini kita ingin menemukan solusi yang berubah halus di dalam
irisan ini dengan mengasumsikan bahwa solusinya bersifat self-similar.
Ini mengimplikasikan bahwa $u$ berbentuk
$$
    u(x, t) = u(\xi),
$$
di mana untuk PDP hiperbolik orde satu variabel self-similar adalah
$$
    \xi = x / t.
$$
Coba masukkan ini ke persamaan asal sehingga menghasilkan solusi dalam
hal $u(\xi)$ untuk bagian dalam irisan.
$$\begin{aligned}
    &u_t + u u_x = 0 \\
    &u' \frac{\partial \xi}{\partial t} + u u' \frac{\partial \xi}{\partial x} = 0 \\
    &u' \left ( -\frac{x}{t^2} + u \frac{1}{t} \right ) = 0 \\
    &u' \left ( -\xi + u\right ) = 0
\end{aligned}$$
Persamaan terakhir ini sekarang dapat diselesaikan.
