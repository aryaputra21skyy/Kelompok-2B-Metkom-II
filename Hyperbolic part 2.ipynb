## Characteristic Tracing

Cara umum untuk menyelesaikan persamaan diferensial parsial (PDE) hiperbolik secara analitik adalah dengan menggunakan *method of characteristics*. Namun, hingga saat ini kita belum benar-benar mencoba menggunakan teori tersebut untuk membangun suatu metode numerik.


Mempertimbangkan nilai solusi pada suatu titik $(x_j, t + \Delta t)$, kita mengetahui bahwa untuk $a > 0$ kita perlu melihat ke belakang menuju titik $(x_j - a \Delta t, t)$, di mana nilai solusi pada titik tersebut menentukan solusi pada titik yang sedang ditinjau.

Pendekatan ini bekerja dengan sangat baik hingga kita mulai mempertimbangkan grid yang terdiskretisasi, di mana kita hanya mengetahui nilai solusi pada waktu $t$ di sejumlah titik diskrit saja.



![alt text](characteristic_tracing_1.png)

Jika karakteristik yang melalui titik $(x_j, t + \Delta t)$ juga melalui suatu titik pada waktu $t$, maka kita tidak mengalami masalah. Namun, biasanya hal ini tidak terjadi kecuali kita secara khusus memilih $\Delta x$ dan $\Delta t$ agar kondisi tersebut terpenuhi. 

Ternyata kondisi ini terjadi jika:
$$
\frac{a \Delta t}{\Delta x} = 1,
$$
yang merupakan batas atas dari hasil kestabilan kita.


![alt text](characteristic_tracing_2.png)

Sebaliknya, jika $\nu < 1$, maka karakteristik tidak akan tepat melalui titik-titik grid. Perhatikan juga bahwa karena adanya batasan $|\nu| \leq 1$, kita mengetahui bahwa karakteristik tidak akan melewati titik $x_{j-1}$.

Sebagai alternatif, kita dapat melakukan interpolasi antara dua nilai yang berada di sekitar lintasan karakteristik tersebut untuk memperkirakan nilai yang dicari. Tunjukkan bahwa jika interpolasi linear digunakan, maka akan diperoleh metode *upwind*.

Untuk interpolasi linear, kita mengetahui bahwa titik perpotongan karakteristik berada di
$$
x_p = x_j - a \Delta t.
$$

Interpolan linear diberikan oleh:
$$
\begin{aligned}
P_1(x) &= \frac{x - x_{j-1}}{x_{j} - x_{j-1}} U^n_{j} + \frac{x - x_j}{x_{j-1} - x_j} U^n_{j-1} \\
       &= \frac{x - x_{j-1}}{\Delta x} U^n_{j} - \frac{x - x_j}{\Delta x} U^n_{j-1}
\end{aligned}
$$

Sehingga nilai solusi pada waktu berikutnya adalah:
$$
\begin{aligned}
U^{n+1}_j = P_1(x_j - a \Delta t) 
&= \frac{x_j - a \Delta t - x_{j-1}}{\Delta x} U^n_{j} 
 - \frac{x_j - a \Delta t - x_j}{\Delta x} U^n_{j-1} \\
&= \frac{\Delta x - a \Delta t}{\Delta x} U^n_{j} 
 + \frac{a \Delta t}{\Delta x} U^n_{j-1} \\
&= U^n_{j} - \frac{a \Delta t}{\Delta x} \left( U^n_{j} - U^n_{j-1} \right).
\end{aligned}
$$

Hasil ini menunjukkan bahwa dengan menggunakan interpolasi linear, kita memperoleh skema numerik:
$$
U^{n+1}_j = U^n_{j} - \nu \left( U^n_{j} - U^n_{j-1} \right),
$$
dengan
$$
\nu = \frac{a \Delta t}{\Delta x}.
$$

Skema ini dikenal sebagai **metode upwind** untuk $a > 0$.

Dengan menggunakan teknik yang serupa, kita juga dapat menurunkan metode **Beam-Warming** dengan interpolasi kuadratik.

Interpolan kuadratik diberikan oleh:
$$
\begin{aligned}
P_2(x) &= \frac{(x - x_{j-1})(x - x_{j-2})}{(x_{j} - x_{j-1}) (x_{j} - x_{j-2})} U^n_{j} \\
&\quad + \frac{(x - x_{j})(x - x_{j-2})}{(x_{j-1} - x_{j}) (x_{j-1} - x_{j-2})} U^n_{j-1} \\
&\quad + \frac{(x - x_{j})(x - x_{j-1})}{(x_{j-2} - x_{j}) (x_{j-2} - x_{j-1})} U^n_{j-2} \\
\\
&= \frac{(x - x_{j-1})(x - x_{j-2})}{2 \Delta x^2} U^n_{j}
 - \frac{(x - x_{j})(x - x_{j-2})}{\Delta x^2} U^n_{j-1}
 + \frac{(x - x_{j})(x - x_{j-1})}{2 \Delta x^2} U^n_{j-2} \\
\\
&= \frac{1}{\Delta x^2} \Bigg[
\frac{1}{2} U^n_{j} (x - x_{j-1})(x - x_{j-2})
- U^n_{j-1} (x - x_{j})(x - x_{j-2})
+ \frac{1}{2} U^n_{j-2} (x - x_{j})(x - x_{j-1})
\Bigg]
\end{aligned}
$$

Kemudian nilai solusi pada waktu berikutnya diperoleh dari:
$$
\begin{aligned}
U^{n+1}_j = P_2(x_j - a \Delta t)
&= \frac{1}{\Delta x^2} \Bigg[
\frac{1}{2} U^n_{j} (\Delta x - a \Delta t)(2 \Delta x - a \Delta t) \\
&\quad - U^n_{j-1} (- a \Delta t)(2 \Delta x - a \Delta t) \\
&\quad + \frac{1}{2} U^n_{j-2} (- a \Delta t)(\Delta x - a \Delta t)
\Bigg] \\
\\
&= \frac{1}{\Delta x^2} \Bigg[
\frac{1}{2} U^n_{j} (2 \Delta x^2 - 3 a \Delta t \Delta x + a^2 \Delta t^2) \\
&\quad - U^n_{j-1} (-2 a \Delta t \Delta x + a^2 \Delta t^2) \\
&\quad + \frac{1}{2} U^n_{j-2} (- a \Delta t \Delta x + a^2 \Delta t^2)
\Bigg] \\
\\
&= U^n_j 
- \frac{a \Delta t}{2 \Delta x} \left(3 U^n_j - 4 U^n_{j-1} + U^n_{j-2}\right) 
+ \frac{a^2 \Delta t^2}{2 \Delta x^2} \left(U^n_j - 2 U^n_{j-1} + U^n_{j-2}\right)
\end{aligned}
$$

Skema ini dikenal sebagai **metode Beam-Warming**, yaitu skema orde dua yang menggunakan pendekatan *upwind* berbasis interpolasi kuadratik.

## Kondisi Courant–Friedrichs–Lewy (CFL)

Salah satu hasil menarik dari analisis karakteristik yang telah kita lakukan adalah bahwa kriteria kestabilan menyebabkan titik perpotongan karakteristik dengan waktu $t_n$ berada dalam interval $[x_{j-1}, x_j]$ ketika $a > 0$. 

Hal ini mencerminkan suatu prinsip yang lebih umum terkait kestabilan dalam metode numerik untuk PDE, yang dikemukakan oleh Courant, Friedrichs, dan Lewy, dan dikenal sebagai **kondisi CFL (CFL condition)**.

Secara lebih umum, untuk PDE lain kita dapat mengharapkan bahwa *domain of dependence* tidak hanya berupa satu titik, melainkan sekumpulan titik (seperti pada sistem persamaan adveksi) atau bahkan suatu interval.

Persamaan panas (*heat equation*) merupakan salah satu contohnya, yang memiliki domain of dependence
$$
\mathcal{D}(X, T) = (-\infty, \infty).
$$
Artinya, semua titik dalam domain mempengaruhi semua titik lainnya pada waktu yang akan datang.

Jenis persamaan seperti ini dikatakan memiliki *kecepatan propagasi tak hingga* (*infinite propagation speed*), yang merupakan karakteristik dari PDE parabolik. Hal ini juga menjadi salah satu cara untuk mengklasifikasikan PDE yang lebih kompleks.

Seseorang mungkin berargumen untuk menolak gagasan ini pada persamaan panas, karena fungsi Green untuk suatu titik tertentu mengalami peluruhan eksponensial saat menjauh dari titik tersebut. Namun, peluruhan ini tetap tidak cukup cepat untuk membatasi pengaruhnya secara lokal.

Inilah yang menjadi dasar kesimpulan sekaligus keterbatasan fisik dari model difusi: material (atau panas) secara matematis dapat menyebar dengan kecepatan tak hingga.


### Domain of Dependence Numerik

Suatu metode numerik juga memiliki *domain of dependence* yang ditentukan oleh *stencil* yang digunakan. 

Sebagai contoh, metode **Lax-Friedrichs** memiliki solusi $U^n_j$ yang bergantung pada titik-titik $U^{n-1}_{j+1}$, $U^{n-1}_{j}$, dan $U^{n-1}_{j-1}$.

Hal ini secara umum berlaku untuk metode tiga titik (*three-point methods*) yang telah kita kembangkan sebelumnya, termasuk metode **Lax-Wendroff**.

Kita juga dapat menelusuri lebih jauh ke belakang dalam waktu untuk melihat titik-titik mana yang mempengaruhi $U^{n-1}_{j+1}$, $U^{n-1}_{j}$, dan $U^{n-1}_{j-1}$. 

Dengan melakukan hal ini, kita akan melihat terbentuknya suatu *cone of dependence* (kerucut ketergantungan) yang semakin melebar seiring kita menelusuri waktu ke belakang.
