# Convergence Results for Initial Value Problems

Konvergensi untuk IVP (Initial Value Problems / masalah nilai awal) sedikit berbeda dibandingkan dengan BVP (Boundary Value Problems / masalah nilai batas). Secara umum, kita menginginkan

$$
    \lim_{\Delta t \rightarrow 0} U^N = 
u(t_f)
$$

di mana $t_f$ adalah waktu akhir yang diinginkan dan N adalah jumlah langkah waktu yang diperlukan untuk mencapai  $t_f$ sehingga berlaku

$$
    N \Delta t = t_f \quad \Rightarrow 
N = \frac{t_f}{\Delta t}.
$$

 Berikut merupakan Salah satu bentuk umum dari IVP:

$$
\frac{dy}{dt} = f(t, y), \quad y(t_0) = y_0
$$

Tujuannya yakni kita  mencari pendekatan numerik dari solusi \( $y(t)$ \) menggunakan metode numerik seperti Euler.

###  Lalu Apasi Itu Konvergensi?
Konvergensi itu artinya mendekati suatu nilai yang benar atau tetap.

sederhananya Dalam matematika/numerik, konvergensi adalah kondisi ketika hasil perhitungan (aproksimasi) makin lama makin dekat dengan nilai sebenarnya.

Sebuah metode numerik dikatakan **konvergen** jika:

$$
\lim_{h \to 0} |y_n - y(t_n)| = 0
$$

Artinya, ketika ukuran langkah \( $h$ \) diperkecil, solusi numerik \( $y_n$ \) makin mendekati solusi eksak \( $y(t_n)$ \).



## One-Step Method Convergence (Konvergensi Metode Satu-Langkah)
adalah metode numerik yang menghitung nilai berikutnya hanya dari satu titik sebelumnya.


Pertimbangkan masalah linear sederhana berikut:

$$    \frac{\text{d}u}{\text{d}t} = \lambda u + g(t) \quad \text{dengan}\quad u(0) = u_0$$

 pertama kita Pisahkan:

$$
\frac{du}{dt} - \lambda u = g(t)
$$

 lalu selesaikan menggunakan faktor integrasi:

$$
\mu(t) = e^{-\lambda t}
$$

Kemudia kali:

$$
e^{-\lambda t}\frac{du}{dt} - \lambda e^{-\lambda t}u = e^{-\lambda t} g(t)
$$

 lalu turunkan :

$$
\frac{d}{dt}\left( e^{-\lambda t} u \right) = e^{-\lambda t} g(t)
$$

 Maka, Integral dari $t_0$ ke $t$:

$$
\int_{t_0}^{t} \frac{d}{dt}\left( e^{-\lambda t} u \right) dt
= \int_{t_0}^{t} e^{-\lambda \tau} g(\tau) d\tau
$$

Hasil:

$$
e^{-\lambda t} u(t) - e^{-\lambda t_0} u_0
= \int_{t_0}^{t} e^{-\lambda \tau} g(\tau) d\tau
$$

selanjutnya kali dengan  $e^{\lambda t}$:

$$
u(t) = u_0 e^{\lambda (t - t_0)}
+ e^{\lambda t} \int_{t_0}^{t} e^{-\lambda \tau} g(\tau) d\tau
$$

Gabungkan eksponen:

$$
u(t) = u_0 e^{\lambda (t - t_0)}
+ \int_{t_0}^{t} e^{\lambda (t - \tau)} g(\tau) d\tau
$$

sehingga hasil akhirnya :

$$    u(t) = u_0 e^{\lambda (t - t_0)} + \int^t_{t_0} e^{\lambda (t - \tau)} g(\tau) d\tau.$$



### Forward Euler on a Linear Problem(Metode Euler Maju pada Masalah Linear)

Misalnya Persamaan Awal
$$
\frac{du}{dt} = \lambda u
$$


Maka dapat diterapkan  metode Euler (Eksplisit/Maju) pada masalah ini 
menghasilkan:
$$\begin{aligned}
U^{n+1} &= U^n + \Delta t\lambda U^n \
&= (1 + \Delta t \lambda) U^n
\end{aligned}$$


Kita juga mengetahui bahwa galat pemotongan lokal (local truncation error) adalah

$$
\tau^n = \frac{u(t_{n+1}) - u(t_n)}{\Delta t} - \lambda u
(t_n)
$$

dimna:
$\tau^n$ = galat lokal pada langkah ke-n

hal ini bisa kita selesaikan dengan:
1. Ekspansi Taylor:
$$
u(t_{n+1}) = u(t_n) + \Delta t u'(t_n) + \frac{1}{2}\Delta t^2 u''(t_n) + \mathcal{O}(\Delta t^3)
$$



2. bagi dengan $\Delta t$:
$$
\frac{u(t_{n+1}) - u(t_n)}{\Delta t}
= u'(t_n) + \frac{1}{2}\Delta t u''(t_n) + \mathcal{O}(\Delta t^2)
$$



3. Substitusi:
$$
\tau^n =
\left(u'(t_n) + \frac{1}{2}\Delta t u''(t_n) + \mathcal{O}(\Delta t^2)\right)
- \lambda u(t_n)
$$

karena:
$$ u'(t_n) = \lambda u(t_n)$$

Maka hasilnya:
$$
\tau^n = \frac{1}{2}\Delta t u''(t_n) + \mathcal{O}(\Delta t^2)
$$

Galat lokal sebanding dengan $Δt$

Dengan memperhatikan definisi awal $\tau^n$, kita dapat 
menulis ulang ekspresi untuk galat pemotongan lokal 
tersebut sebagai:

$$
\frac{u(t_{n+1}) - u(t_n)}{\Delta t}
= \lambda u(t_n) + \tau^n
$$

Dikalikan $ \Delta t$:
$$
u(t_{n+1}) = u(t_n) + \Delta t \lambda u(t_n) + \Delta t 
\tau^n
$$


sehingga:
$$
u(t_{n+1}) = (1 + \Delta t \lambda) u(t_n) + \Delta t \tau^n
$$


yang kemudian, jika digabungkan dengan penerapan metode Euler, menghasilkan suatu ekspresi untuk galat global:

$$\begin{aligned}
E^{n+1} = U^{n+1} - u(t^{n+1}) &= (1 + \Delta t \lambda) 
U^n - (1 + \Delta t \lambda) u(t_n) - \Delta t \tau^n \\
&= (1+\Delta t \lambda) E^n - \Delta t \tau^n \\
\end{aligned}$$

Dengan mengembangkan (mengekspansi) persamaan ini ke belakang hingga langkah awal $n=0$ diperoleh:$$    E^n = (1 + \Delta t \lambda)^n E^0 - \Delta t \sum^n_{i=1} (1 + \Delta t \lambda)^{n-i} \tau^{i - 1}.$$

Sekarang kita dapat melihat pentingnya suku 
$(1 + \Delta t \lambda)$. Suku ini dapat dibatasi (dibound) oleh::
$$    |1 + \Delta t \lambda| \leq e^{\Delta t |\lambda|}$$

yang kemudian mengimplikasikan bahwa suku dalam penjumlahan dapat dibatasi oleh:$$    |1 + \Delta t \lambda|^{n - i} \leq e^{(n-i) \Delta t |\lambda|} \leq e^{n \Delta t |\lambda|} \leq e^{|\lambda| t_f}$$

Dengan menggunakan hasil pembatasan tersebut ke dalam ekspresi galat global, kita peroleh:
$$\begin{aligned}
    E^n &= (1 + \Delta t \lambda) E^0 - \Delta t \sum^n_
{i=1} (1 + \Delta t \lambda)^{n-i} \tau^{i - 1} \\
    |E^n| &\leq e^{|\lambda| \Delta t} |E^0| - \Delta t 
\sum^n_{i=1} e^{|\lambda| t_f} |\tau^{i - 1}| \\
          &\leq e^{|\lambda| t_f} \left(|E^0| - \Delta t 
\sum^n_{i=1} |\tau^{i - 1}|\right) \\
          &\leq e^{|\lambda| t_f} \left(|E^0| - n \Delta t
\max_{1 \leq i \leq n} |\tau^{i - 1}|\right)
\end{aligned}$$

Dengan kata lain, galat global dibatasi oleh galat awal dan galat maksimum satu langkah yang dikalikan dengan jumlah langkah waktu yang diambil. Jika $N
= \frac{t_f}{\Delta t}$ seperti sebelumnya dan dengan 
seperti sebelumnya, dan dengan mempertimbangkan galat pemotongan lokal, maka kita dapat menyederhanakan ekspresi ini menjadi:

$$    |E^n| \leq e^{|\lambda| t_f} \left[|E^0| + t_f \left
(\frac{1}{2} \Delta t |u''| + \mathcal{O}(\Delta t^2)
\right ) \right]$$

Jika kita mengasumsikan bahwa kita telah menggunakan kondisi awal $u_0$ yang benar, maka $E^0 \rightarrow 0$ seiring dengan $\Delta t \rightarrow 0$. Dengan demikian, kita dapat melihat bahwa metode tersebut benar-benar konvergen karena:
$$    |E^n| \leq e^{|\lambda| t_f} t_f \left(\frac{1}{2} \Delta t |u''| + \mathcal{O}(\Delta t^2)\right ) = \mathcal{O}(\Delta t).$$

### Relation to Stability for BVPs (Hubungan dengan Stabilitas untuk BVP)
Kita dapat melihat hubungan antara definisi stabilitas sebelumnya dengan yang telah dijelaskan di atas. Cobalah menuliskan metode Euler maju (Forward Euler) sebagai suatu sistem linear.

Euler Eksplisit (Forward Euler):
$$
    A = \frac{1}{\Delta t} \begin{bmatrix}
        1 \\
        -(1 + \Delta t \lambda) & 1 \\
        & -(1 + \Delta t \lambda) & 1 \\
        & & -(1 + \Delta t \lambda) & 1 \\
        & & & \ddots & \ddots \\
        & & & & -(1 + \Delta t \lambda) & 1 \\
        & & & & & -(1 + \Delta t \lambda) & 1
    \end{bmatrix}
$$
dengan 
$$
    U = \begin{bmatrix} U^1 \\ U^2 \\ \vdots \\ U^N \end{bmatrix} ~~~~ 
    F = \begin{bmatrix} (1 / \Delta t + \lambda) U^0 + g(t_0) \\ g(t_1) \\ \vdots \\ g(t_{N-1}) \end{bmatrix}
$$

Mengikuti hasil stabilitas kita sebelumnya dan dengan menetapkan $\hat{U}$ sebagai vektor yang diperoleh dari solusi eksak ($\hat{U}^i = u(t_i)$), maka kita memiliki:
$$    A U = F  ~~~~~~ A \hat{U} = F + \tau$$

 maka dari  itu,$$    A (\hat{U} - U) = \tau.$$

Dengan memperhatikan bahwa $\hat{U~} - U = E$ , kita kemudian dapat membalik (invert) matriks  $A$ untuk menemukan hubungan antara galat pemotongan $\tau$ dan galat global E.

Seperti sebelumnya, kita memerlukan bahwa $A^{-1}$
dapat diinvers (yang dalam kasus ini bersifat sepele/mudah) dan bahwa  $||A^{-1}|| < C$

Hal ini dapat kita lihat sebagai berikut.

$$
    A^{-1} = \Delta t \begin{bmatrix}
        1 \\
        (1 + \Delta t \lambda) & 1 \\
        (1 + \Delta t \lambda)^2 & (1 + \Delta t \lambda) & 1 \\
        (1 + \Delta t \lambda)^3 & (1 + \Delta t \lambda)^2 & (1 + \Delta t \lambda) & 1\\
        \vdots & & & \ddots \\
        (1 + \Delta t \lambda)^{N-1} & (1 + \Delta t \lambda)^{N-2} (1 + \Delta t \lambda)^{N-3} & \cdots & (1 + \Delta t \lambda) & 1 
   \end{bmatrix}
$$
yang norma tak hingganya (infinity norm) adalah
$$
    ||A^{-1}||_\infty = \Delta t \sum^N_{m=1} | (1 + \Delta t \lambda)^{N-M} |
$$
dan oleh karena itu
$$
    ||A^{-1}||_\infty \leq \Delta t N e^{|\lambda| T} = T e^{|\lambda| T}.
$$
As $\Delta t \rightarrow 0$ ini bernilai terbatas untuk $T$ yang tetap (fixed $T$).

### General One-Step Method Convergence (Konvergensi Metode Satu-Langkah Umum)

Dengan asumsi bahwa $\Phi$ kontinu dalam $t$ dan $\Delta t$, serta kontinu Lipschitz dalam $u$ dengan konstanta Lipschitz $L$ (yang berkaitan dengan konstanta Lipschitz dari fungsi $f$). Jika metode satu-langkah tersebut konsisten:
$$    \Phi(u,t,0) = f(u,t)$$
untuk semua $u$, $t$, dan $\Delta t$, dan galat pemotongan lokalnya (local truncation error) adalah:
$$    \tau^n =\frac{u(t_{n+1}) - u(t_n)}{\Delta t} - \Phi(u(t_n), t_n, \Delta t)$$


maka metode satu-langkah tersebut konvergen.

Menggunakan pendekatan umum yang kita gunakan untuk metode Euler eksplisit, kita mengetahui bahwa solusi eksak dan $\tau$ dihubungkan melalui:

$$
    u(t_{n+1}) = u(t_n) + \Delta t \Phi(u(t_n), t_n, \Delta t) + \Delta t \tau^n
$$
yang jika dikurangkan dari solusi pendekatan (approximate solution):
$$
    U^{n+1} = U^n + \Delta t \Phi(U^n, t_n, \Delta t)
$$
menghasilkan:
$$
    E^{n+1} = E^n + \Delta t (\Phi(U^n, t_n, \Delta t) - \Phi(u(t_n), t_n, \Delta t)) - \Delta t \tau^n.
$$

Penyederhanaan
$$
E^{n+1}
=
(U^n - u(t_n))
+ \Delta t \big(\Phi(U^n, t_n, \Delta t) - \Phi(u(t_n), t_n, \Delta t)\big)
- \Delta t \tau^n
$$

$$
E^{n+1}
=
E^n
+ \Delta t \big(\Phi(U^n, t_n, \Delta t) - \Phi(u(t_n), t_n, \Delta t)\big)
- \Delta t \tau^n
$$

dengan menggunakan Sifat Lipschitz dari $\Phi$:
$$
|\Phi(U^n, t_n, \Delta t) - \Phi(u(t_n), t_n, \Delta t)|
\le L |U^n - u(t_n)|
$$

$$
= L |E^n|
$$

 kita kemudian memperoleh:
$$    |E^{n+1}| \leq |E^n| + \Delta t L |E^n| + \Delta t |\tau^n|.$$
yang memiliki bentuk yang sama dengan yang kita lihat pada pembuktian metode Euler eksplisit (forward Euler).

## Zero-Stability for Linear Multistep Methods (Kestabilan Nol untuk Metode-Metode Multilangkah Linier)
Kita juga bisa membuat pernyataan umum untuk metode-metode 
multilangkah linier meskipun penting untuk dicatat bahwa 
kita memiliki syarat-syarat tambahan untuk metode-metode 
multilangkah linier agar konvergen.  Sebagai contoh, 
pertimbangkan metode
$$
 U^{n+2} - 3 U^{n+1} + 2 U^n = - \Delta t f(U^n)
$$
sehingga $\alpha_0 = 2$, $\alpha_1 = -3$, dan $\alpha_2 = 
1$ serta $\beta_0 = -1$ dengan sisanya sama dengan nol.  
Perhatikan bahwa koefisien-koefisien ini memenuhi syarat 
untuk konsisten dengan kesalahan pemotongan
$$
 \tau^n = \frac{1}{\Delta t} (u(t_{n+2}) - 3 u(t_{n+1}) + 2 
u(t_n) + \Delta t u'(t_n)) = \frac{5}{2} \Delta t u''(t_n) 
+ \mathcal{O}(\Delta t^2).
$$
Ternyata meskipun metode ini konsisten, kesalahan global 
tidak konvergen secara umum!

Pada kasus ODE trivial $u'(t) = 0$, solusi eksaknya adalah $u(t) = 0$. Jika metode multilangkah linier yang digunakan tidak zero-stable, maka galat kecil pada data awal bisa membesar secara eksponensial.

Pada contoh: $$ U^{n+2} - 3 U^{n+1} + 2 U^n = 0 $$ Solusi umumnya: $$ U^n = 2 U^0 - U^1 + 2^n (U^1 - U^0) $$ Jika $U^1$ tidak tepat (ada galat kecil), maka suku $2^n (U^1 - U^0)$ akan membesar sangat cepat seiring $n$ bertambah, meskipun galat awal sangat kecil.

Jika kita tidak memiliki nilai awal $U^1$ yang tepat (misal, ada sedikit galat pada $U^1$), maka solusi umum dari persamaan beda linier tersebut adalah:

$$ U^n = 2 U^0 - U^1 + 2^n (U^1 - U^0) $$

Perhatikan bahwa terdapat suku $2^n (U^1 - U^0)$. Jika $U^1 - U^0$ mengandung galat kecil, suku ini akan membesar sangat cepat seiring $n$ bertambah, karena dipangkatkan dengan $2^n$.

Artinya, galat kecil pada data awal akan membesar secara eksponensial. Inilah ciri khas metode yang tidak zero-stable: galat awal tidak terkontrol dan solusi numerik menjadi tidak berguna, bahkan jika metode tersebut konsisten.


### Characteristic Polynomials and Linear Difference Equations(Polinomial Karakteristik dan Persamaan Beda Linier)

Untuk memahami kestabilan dari metode numerik, khususnya Linear Multistep Method (LMM), kita mulai dengan mempertimbangkan bentuk umum dari persamaan beda linier berikut:

$$\sum^r_{j=0} \alpha_j U^{n+j} = 0$$

dengan $U^0, U^1, \ldots, U^{r-1}$ sebagai kondisi awal, Kita asumsikan solusi umum berbentuk eksponensial:
$$U^n = \xi^n$$

Jika kita substitusikan bentuk ini ke dalam persamaan beda di atas, kita memperoleh:
$$
 \sum^r_{j=0} \alpha_j \xi^{n+j} = 0
$$

Dengan membagi semua persamaan dengan $\xi^n$ (dengan asumsi $\xi \neq 0$), kita peroleh:
$$
 \sum^r_{j=0} \alpha_j \xi^j = 0 
$$
Persamaan ini disebut polinomial karakteristik 
$$
 \rho(\xi) = \sum^r_{j=0} \alpha_j \xi^j
$$
Dengan demikian, nilai $\xi$ yang memenuhi persamaan tersebut merupakan akar-akar dari polinomial karakteristik tersebut. Karena metode ini bersifat linier, maka kombinasi linier dari solusi-solusi juga merupakan solusi. Oleh karena itu, bentuk umum dari solusi adalah:
$$
 U^n = c_1 \xi_1^n + c_2 \xi_2^n + \cdots + c_r \xi^n_r.
$$
Dengan kondisi awal $U^0, U^1, \ldots, U^{r-1}$, kita dapat menentukan nilai-nilai konstanta $c_j$ secara unik.

###  General Zero-Stability Result for LMM (Hasil Umum Kestabilan Nol untuk LMM)
Sebuah metode multilangkah linier dikatakan zero-stable jika semua akar dari polinomial karakteristik $\rho(\xi)$ memenuhi:

1. $|\xi_j| \leq 1$ untuk semua $j = 1,2,\ldots,r$, dan

2. jika $|\xi_j| = 1$, maka $\xi_j$ tidak boleh berulang (tidak boleh menjadi akar ganda).


### Contoh : Metode Multilangkah
Perhatikan metode multilangkah berikut:
$$
 U^{n+2} - 2 U^{n+1} + U^n = \frac{\Delta t}{2} (f(U^{n+2}) - f(U^n)).
$$
Jika kita mengujinya pada ODE sederhana $u'(t) = 0$, maka sisi kanan menjadi nol:
$$
 U^{n+2} - 2 U^{n+1} + U^n = 0
$$
Polinomial karakteristiknya adalah:
$$
 \rho(\xi) = \xi^2 - 2 \xi + 1 = (\xi - 1)^2
$$
yang memiliki akar ganda di $\xi = 1$.
Solusi umumnya menjadi:
$$
 U^n = c_1 + c_2 n.
$$

Artinya, meskipun kondisi awal kecil, solusi dapat meningkat secara linier terhadap waktu. Karena ada akar ganda pada $|\xi| = 1$, maka metode ini tidak zero-stable.


### Absolute Stability (Stabilitas Mutlak)

Meskipun stabilitas nol menjamin stabilitas, hal ini jauh lebih sulit untuk digunakan secara umum karena batas $\Delta t \rightarrow 0$ bisa jadi sulit untuk dihitung.  Sebagai gantinya, kita sering mempertimbangkan sebuah $\Delta t\ yang terbatas dan memeriksa apakah metode ini stabil untuk pilihan $\Delta t\ ini.  Hal ini memiliki keuntungan praktis karena akan memberitahu kita $\Delta t$ tertentu yang akan memastikan bahwa metode kita memang stabil.

### Example
Pertimbangkan masalah ini

$$u'(t) = \lambda (u - \cos t) - \sin t \quad \text{dengan} \quad u(0) = 1$$

yang solusi tepatnya adalah 

$$u(t) = \cos t.$$
Kita dapat menghitung perkiraan untuk $\Delta t$ yang harus kita gunakan dengan memeriksa kesalahan pemotongan 

$$\begin{aligned}
    \tau &= \frac{1}{2} \Delta t u''(t) + \mathcal{O}(\Delta t^2) \\
 &= -\frac{1}{2} \Delta t \cos t + \mathcal{O}(\Delta t^2)
\end{aligned}$$

dan oleh karena itu

$$|E^n| \leq \Delta t \max_{0 \leq t \leq t_f} |\cos t| = \Delta t.$$

Jika kita menginginkan sebuah solusi di mana $|E^n| < 10^{-3}$ maka $\Delta t \approx 10^{-3}$.  Beralih ke penerapan metode Euler, mari kita terapkan ini pada kasus di mana $\lambda = -10$ dan $\lambda = -2100$.

```python
import numpy as np                  # Library untuk operasi numerik (array, fungsi matematika)
import matplotlib.pyplot as plt     # Library untuk plotting grafik

plt.style.use('default')            # Reset style supaya warna default (tidak abu-abu/ggplot)

# =========================
# DEFINISI FUNGSI MASALAH
# =========================

# Fungsi ODE: du/dt = λ(u - cos(t)) - sin(t)
f = lambda t, lam, u: lam * (u - np.cos(t)) - np.sin(t)

# Solusi eksak (analitik)
u_exact = lambda t: np.cos(t)

# =========================
# PARAMETER SIMULASI
# =========================

t_f = 2.0                           # Waktu akhir simulasi
num_steps = [2**n for n in range(4, 10)]  # Jumlah grid (16, 32, 64, ...)

# Array untuk menyimpan hasil
delta_t = np.empty(len(num_steps))     # Menyimpan ukuran langkah waktu
error_10 = np.empty(len(num_steps))    # Error untuk λ = -10
error_2100 = np.empty(len(num_steps))  # Error untuk λ = -2100 (stiff)

# =========================
# PERHITUNGAN NUMERIK (EULER)
# =========================

for (i, N) in enumerate(num_steps):
    
    t = np.linspace(0, t_f, N)        # Membuat grid waktu dari 0 ke t_f
    delta_t[i] = t[1] - t[0]          # Menghitung Δt
    
    # ===== Kasus 1: λ = -10 (tidak stiff) =====
    U = np.empty(t.shape)             # Array solusi numerik
    U[0] = 1.0                        # Kondisi awal
    
    # Loop metode Euler
    for (n, t_n) in enumerate(t[1:]):
        U[n+1] = U[n] + delta_t[i] * f(t_n, -10.0, U[n])
    
    # Hitung error relatif di waktu akhir
    error_10[i] = np.abs(U[-1] - u_exact(t_f)) / np.abs(u_exact(t_f))
    
    # ===== Kasus 2: λ = -2100 (stiff problem) =====
    U = np.empty(t.shape)
    U[0] = 1.0
    
    for (n, t_n) in enumerate(t[1:]):
        U[n+1] = U[n] + delta_t[i] * f(t_n, -2100.0, U[n])
    
    error_2100[i] = np.abs(U[-1] - u_exact(t_f)) / np.abs(u_exact(t_f))

# =========================
# FUNGSI GARIS REFERENSI ORDE 1
# =========================

# Digunakan untuk membuat garis slope orde 1 (Δt^1)
order_C = lambda delta_x, error, order: np.exp(np.log(error) - order * np.log(delta_x))

# =========================
# PLOTTING
# =========================

fig = plt.figure(figsize=(12,5))   # Membuat figure (lebar diperbesar)

# ===== Grafik kiri (λ = -10) =====
axes1 = fig.add_subplot(1, 2, 1)   # Subplot 1 baris, 2 kolom, posisi 1

# Plot error numerik
axes1.loglog(delta_t, error_10,
             color='blue',         # warna biru
             marker='o',           # marker bulat
             linestyle='None',     # tanpa garis
             label='Forward Euler')

# Plot garis referensi orde 1
axes1.loglog(delta_t,
             order_C(delta_t[1], error_10[1], 1.0) * delta_t,
             color='red',
             linestyle='--',
             linewidth=2,
             label='1st Order')

# Label & tampilan
axes1.set_title("Comparison of Errors (λ = -10)")
axes1.set_xlabel("$\\Delta t$")
axes1.set_ylabel("$|U(t_f) - u(t_f)|$")
axes1.grid(True, which='both', linestyle='--', linewidth=0.5)
axes1.legend()

# ===== Grafik kanan (λ = -2100) =====
axes2 = fig.add_subplot(1, 2, 2)

axes2.loglog(delta_t, error_2100,
             color='green',
             marker='o',
             linestyle='None',
             label='Forward Euler')

axes2.loglog(delta_t,
             order_C(delta_t[1], error_2100[1], 1.0) * delta_t,
             color='black',
             linestyle='--',
             linewidth=2,
             label='1st Order')

axes2.set_title("Comparison of Errors (λ = -2100)")
axes2.set_xlabel("$\\Delta t$")
axes2.set_ylabel("$|U(t_f) - u(t_f)|$")
axes2.grid(True, which='both', linestyle='--', linewidth=0.5)
axes2.legend()

# Mengatur layout agar tidak saling tumpang tindih
plt.tight_layout()

# Menampilkan grafik
plt.show()
```

Jadi, apa yang salah dengan $\lambda = -2100$?  Kesalahan global seharusnya dituliskan sebagai

$$E^{n+1} = (1 + \Delta t \lambda) E^n - \Delta t T^n$$

Jika $\Delta t \approx lebih 10^{-3}$ maka untuk kasus $\lambda = -10$ kesalahan global sebelumnya dikalikan dengan

$$1 + 10^{-3} \cdot -10 = 0.99$$

yang berarti kontribusi dari $E^n$ perlahan-lahan akan berkurang seiring dengan bertambahnya langkah waktu.  Untuk kasus lainnya, kita memiliki

$$1 + 10^{-3} \cdot -2100 = -1.1$$

yang berarti bahwa untuk $\Delta t$ ini, kesalahan yang dibuat pada langkah waktu sebelumnya akan bertambah!  Agar hal ini tidak terjadi, kita harus memiliki $\Delta t < 1 / 2100$ yang akan menyebabkan konvergensi lagi.


### Absolute Stability of the Forward Euler Method (Kestabilan Mutlak dari Metode Euler Maju)


Pertimbangkan kembali soal uji sederhana $u'(t) 
= \lambda u$.  Kita sudah mengetahui sebelumnya 
bahwa menerapkan metode Euler pada masalah ini 
akan menghasilkan sebuah pembaruan dalam bentuk


$$U_{n+1} = (1 + \Delta t \lambda) U_n.$$


Seperti yang sudah dijelaskan pada contoh 
sebelumnya, kita mengetahui bahwa jika


$$|1 + \Delta t \lambda| \leq 1$$


metode ini akan menjadi stabil, hal ini 
dinamakan **kestabilan absolut**.  Perhatikan 
bahwa hasil kali dari $\Delta t \lambda$ adalah 
hal yang penting di sini dan sering kita anggap 
sebagai **daerah stabilitas absolut** pada 
dataran kompleks yang didefinisikan oleh 
persamaan yang diuraikan di mana sekarang $z = 
\Delta t \lambda$.  Hal ini memungkinkan nilai 
$\lambda$ menjadi kompleks yang dapat menjadi 
kasus penting untuk dipertimbangkan, terutama 
untuk sistem persamaan di mana $\lambda$ 
diidentifikasi sebagai nilai eigen.

### General Stability Regions for Linear Multistep Methods (Daerah Stabilitas Umum untuk Metode Linear Multistep)

Kembali ke metode linear multistep dan menerapkannya secara umum pada masalah pengujian kita, kita memiliki
$$
\sum^r_{j=0} \alpha_j U_{n+j} = \Delta t \sum^r_{j=0} \beta_j \lambda U_{n+j}
$$
yang dapat ditulis sebagai
$$
\sum^r_{j=0} (\alpha_j - \beta_j \Delta t \lambda) U_{n+j} = 0
$$
atau menggunakan notasi kita $z = \Delta t \lambda$ kita memiliki
$$
\sum^r_{j=0} (\alpha_j - \beta_j z) U_{n+j} = 0.
$$

Ini memiliki bentuk yang mirip dengan persamaan perbedaan linear yang dipertimbangkan di atas! Misalkan
$$
\rho(\xi) = \sum^r_{j=0} \alpha_j \xi^j
$$
dan
$$
\sigma(\xi) = \sum^r_{j=0} \beta_j \xi^j
$$
kita dapat menulis ekspresi di atas sebagai
$$
\pi(\xi, z) = \rho(\xi) - z \sigma(\xi)
$$
disebut **polinomial stabilitas** dari metode multi-langkah linear.

Ternyata jika akar $\xi_i$ dari polinomial ini memenuhi
$$
|\xi_i| \leq 1
$$
maka metode multi-langkah tersebut benar-benar stabil. Kita kemudian mendefinisikan wilayah stabilitas absolut sebagai nilai untuk $z$ yang memenuhi syarat ini. Pendekatan ini juga dapat diterapkan pada metode satu-langkah.

### Contoh: Metode Euler Maju

Dengan meneliti metode Euler maju, kita memiliki

$$\begin{aligned}
0 &= U_{n+1} - U_n - \Delta t \lambda U_n \\
&= U_{n+1} - U_n (1 + \Delta t \lambda)\\
&= \xi - 1 (1 + z)\\
&=\pi(\xi, z)
\end{aligned}$$

yang akarnya adalah $\xi = 1 + z$ dan kita telah menurunkan kembali daerah stabilitas yang telah kita temukan sebelumnya.

### Absolute Stability of the backward Euler Method (Stabilitas Absolut Metode Euler Terbalik)


Versi terbalik dari metode Euler didefinisikan sebagai
$$
U_{n+1} = U_n + \Delta t f(t_{n+1}, U_{n+1}).
$$
Periksa untuk melihat apakah Euler terbalik stabil secara absolut.
Jika kita pertimbangkan lagi masalah pengujian dari sebelumnya, kita menemukan bahwa
$$\begin{aligned}
0 &= U_{n+1} (1 - \Delta t \lambda) - U_n \\
&= \xi (1 - z) - 1
\end{aligned}$$
yang memiliki akar $\xi = \frac{1}{1 - z}$. Kita kemudian memiliki
$$
\left|\frac{1}{1-z}\right| \leq 1 \leftrightarrow |1 - z| \geq 1
$$
sehingga pada kenyataannya daerah stabilitas mencakup seluruh bidang kompleks kecuali lingkaran yang berpusat di $(1, 0)$ dengan jari-jari 1 yang menyiratkan bahwa metode Euler terbalik pada kenyataannya stabil untuk setiap pilihan $\Delta t$.

```python
import numpy as np                  # Library numerik (exp, cos, array, dll)
import matplotlib.pyplot as plt     # Library untuk plotting grafik

# =========================================
# DEFINISI SOLUSI ODE
# =========================================

# Solusi umum ODE:
# u(t) = e^{λ(t - t0)} (η - cos(t0)) + cos(t)
# t0  = waktu awal
# eta = kondisi awal u(t0)
# lam = parameter λ
u = lambda t_0, eta, lam, t: np.exp(lam * (t - t_0)) * (eta - np.cos(t_0)) + np.cos(t)

# =========================================
# PARAMETER SIMULASI
# =========================================

num_steps = 10                          # Jumlah langkah waktu (dibagi-bagi interval)
error = np.ones(num_steps) * 1.0        # Besar error (gangguan) tiap langkah
t_hat = np.linspace(0.0, 10.0, num_steps + 1)  
# Titik-titik waktu diskrit (grid kasar)

t_whole = np.linspace(0.0, 10.0, 1000) 
# Waktu kontinu untuk plot garis referensi

# =========================================
# SETUP PLOT
# =========================================

fig = plt.figure()                      # Membuat figure
axes = fig.add_subplot(1, 1, 1)         # 1 subplot

eta = 1.0                               # Kondisi awal pertama
lam = 0.1                               # Nilai λ (di sini positif → cenderung tidak stabil)

# =========================================
# LOOP PER LANGKAH WAKTU
# =========================================

for n in range(1, num_steps):
    
    # Waktu lokal dari langkah ke-(n-1) ke n
    t = np.linspace(t_hat[n-1], t_hat[n], 100)
    
    # Solusi ODE pada interval ini
    U = u(t_hat[n-1], eta, lam, t)
    
    # Plot solusi lokal (biru solid)
    axes.plot(t, U, 'b')
    
    # Plot solusi "ideal" dari titik awal itu (biru putus-putus)
    axes.plot(t_whole, u(t_hat[n-1], eta, lam, t_whole), 'b--')
    
    # =========================================
    # MENAMBAHKAN ERROR (GANGGUAN)
    # =========================================
    
    axes.plot([t[-1], t[-1]], 
              (U[-1], U[-1] + -1.0**n * error[n]), 
              'r')
    # Garis merah vertikal = gangguan/error pada akhir langkah
    
    # Update kondisi awal untuk langkah berikutnya
    eta = U[-1] + -1.0**n * error[n]
    # Error ditambahkan bergantian (+ dan -)

# =========================================
# PLOT SOLUSI ASLI
# =========================================

t = np.linspace(0.0, 10.0, 100)
axes.plot(t, np.cos(t), 'g')  
# Garis hijau = solusi eksak cos(t)

# =========================================
# LABEL DAN TAMPILAN
# =========================================

axes.set_title("Perturbed Solutions $\lambda = %s$" % lam)
axes.set_xlabel('$t$')
axes.set_ylabel('$u(t)$')

axes.set_ylim((-10,10))   # Batas sumbu y supaya semua terlihat jelas

# =========================================
# TAMPILKAN GRAFIK
# =========================================

plt.show()
```

### Contoh: Sistem kimia

Pertimbangkan transisi zat kimia $A$ ke zat kimia $C$ melalui proses

$$A \overset{K_1}{\rightarrow} B \overset{K_2}{\rightarrow} C.$$

Jika kita biarkan

$$\vec{u} = \begin{bmatrix} [A] \\ [B] \\ [C] \end{bmatrix}$$

maka kita dapat memodelkan reaksi kimia sederhana ini dengan sistem ODE

$$\frac{\text{d} \vec{u}}{\text{d} t} =
\begin{bmatrix}

-K_1 & 0 & 0 \\
K_1 & -K_2 & 0 \\
0 & K_2 & 0
\end{bmatrix} \vec{u}$$

Solusi sistem ini adalah bentuk

$$u_j(t) = c_{j1} e^{-K_1 t} + c_{j2} e^{-K_2 t} + c_{j3}$$

```python
import numpy as np                  # Library numerik (array, matriks, dll)
import matplotlib.pyplot as plt     # Library untuk plotting grafik

# =========================================
# PARAMETER SISTEM KIMIA
# =========================================

K_1 = 3        # Konstanta reaksi A → B
K_2 = 1        # Konstanta reaksi B → C

# Matriks sistem (model reaksi berantai A → B → C)
A = np.array([[-K_1, 0,    0],     # d[A]/dt = -K1*A
              [ K_1, -K_2, 0],     # d[B]/dt = K1*A - K2*B
              [ 0,    K_2, 0]])    # d[C]/dt = K2*B

# Fungsi ODE sistem (vector form)
f = lambda u: np.dot(A, u)         # f(u) = A * u

# =========================================
# DISKRETISASI WAKTU
# =========================================

t = np.linspace(0.0, 8.0, 128)     # Waktu dari 0 sampai 8 (128 titik)
delta_t = t[1] - t[0]              # Langkah waktu Δt

# =========================================
# INISIALISASI SOLUSI
# =========================================

U = np.empty((t.shape[0], 3))      # Matriks solusi (waktu x 3 variabel)
U[0, :] = [2.5, 5.0, 2.0]          # Kondisi awal: [A], [B], [C]

# =========================================
# METODE EULER MAJU
# =========================================

for n in range(t.shape[0] - 1):
    U[n+1, :] = U[n, :] + delta_t * f(U[n, :])
    # Rumus Euler:
    # U(n+1) = U(n) + Δt * f(U(n))

# =========================================
# PLOTTING
# =========================================

fig = plt.figure()                 
axes = fig.add_subplot(1, 1, 1)    # 1 subplot

axes.plot(t, U)                    # Plot semua kolom U (A, B, C) terhadap waktu

# =========================================
# LABEL DAN TAMPILAN
# =========================================

axes.set_title("Chemical System")  # Judul grafik
axes.set_xlabel("$t$")             # Label sumbu x (waktu)

axes.set_title("$[A], [B], [C]$")  # (NOTE: ini overwrite judul sebelumnya)
# ⚠️ ini sebenarnya menimpa title sebelumnya, jadi hanya ini yang tampil

axes.set_ylim((0.0, 10.))          # Batas sumbu y (konsentrasi)

# =========================================
# TAMPILKAN GRAFIK
# =========================================

plt.show()
```

### Apa itu kekakuan?

Secara umum, ODE **kaku** adalah ODE di mana $u'(t) \ll f'(t, u)$. Untuk sistem ODE, **rasio kekakuan**

$$\frac{\max_p |\lambda_p|}{\min_p |\lambda_p|}$$

dapat digunakan untuk mengkarakterisasi kekakuan sistem. Dalam contoh terakhir, rasio ini adalah $K_1 / K_2$ jika $K_1 > K_2$. Saat kami meningkatkan rasio ini, kami mengamati bahwa metode numerik menjadi tidak stabil, hanya pengurangan $\Delta t$ yang menghasilkan solusi stabil lagi. Untuk metode langkah waktu eksplisit, ini bermasalah karena pengurangan langkah waktu hanya untuk satu spesies menyebabkan evaluasi yang sangat mahal. Misalnya, Euler maju memiliki kriteria stabilitas

$$|1 + \Delta t \lambda| < 1$$

di mana $\lambda$ harus menjadi nilai eigen maksimum dari sistem.

```python
import numpy as np                  # Library numerik (array, cos, sin, dll)
import matplotlib.pyplot as plt     # Library untuk plotting grafik

# =========================================
# MEMBUAT FIGURE DAN AXES
# =========================================

fig = plt.figure()                  # Membuat canvas utama
axes = fig.add_subplot(1, 1, 1)     # Membuat 1 subplot

# =========================================
# MEMBUAT PARAMETER LINGKARAN
# =========================================

t = np.linspace(0.0, 2.0 * np.pi, 100)  
# 100 titik dari 0 sampai 2π → untuk menggambar lingkaran

# =========================================
# PARAMETER SISTEM
# =========================================

K_1 = 3.0        # Konstanta reaksi pertama
K_2 = 1.0        # Konstanta reaksi kedua
delta_t = 1.0    # Langkah waktu

# Eigenvalue sistem (dari matriks A)
eigenvalues = [-K_1, -K_2]  
# Untuk sistem kimia: λ = -K1 dan -K2

# =========================================
# MENGGAMBAR DAERAH STABILITAS
# =========================================

axes.fill(np.cos(t) - 1.0, np.sin(t), color='lightgray')  
# Menggambar daerah stabilitas Euler:
# |1 + z| ≤ 1 → lingkaran radius 1, pusat (-1, 0)

# =========================================
# MEM-PLOT TITIK EIGENVALUE
# =========================================

for lam in eigenvalues:
    print(lam * delta_t)            # Menampilkan nilai λΔt di console
    
    axes.plot(lam * delta_t, 0.0, 'ko')  
    # Plot titik (λΔt, 0) sebagai titik hitam (ko)
    # Ini menunjukkan posisi eigenvalue di bidang kompleks

# =========================================
# MENGGAMBAR SUMBU X DAN Y
# =========================================

axes.plot([-3, 3], [0.0, 0.0], 'k--')  
# Garis horizontal (sumbu x)

axes.plot([0.0, 0.0], [-3, 3], 'k--')  
# Garis vertikal (sumbu y)

# =========================================
# PENGATURAN TAMPILAN
# =========================================

# axes.set_xlim((-3, 1))            # (opsional) batas x
axes.set_ylim((-2, 2))              # Batas sumbu y

axes.set_aspect('equal')            # Skala x dan y sama (biar lingkaran tidak oval)

# =========================================
# JUDUL
# =========================================

axes.set_title("Absolute Stability Region for Forward Euler")

# =========================================
# TAMPILKAN GRAFIK
# =========================================

plt.show()
```

### A-Stabilitas
Bagaimana jika kita dapat memperluas wilayah stabilitas absolut untuk mencakup lebih banyak bidang setengah kiri atau bahkan lebih baik, semuanya. Metode yang memiliki properti ini disebut **A-stabil**. Kita telah melihat satu contohnya dengan Euler terbalik yang memiliki wilayah stabilitas

$$|1 - z| \geq 1$$

yang mencakup seluruh bidang setengah kiri.

Ternyata untuk metode multi-langkah linear, teorema Dahlquist membuktikan bahwa tidak ada LMM yang memenuhi kriteria stabilitas A di luar orde kedua (aturan trapesium). Akan tetapi, ada metode Runge-Kutta orde lebih tinggi.

Mungkin ini terlalu membatasi. Seringkali nilai eigen yang besar untuk sistem (misalnya yang berasal dari diskritisasi PDE untuk persamaan panas) terletak sepenuhnya pada garis riil. Jika daerah stabilitas dapat mencakup sebanyak mungkin garis riil sambil mengabaikan sisa bidang setengah kiri, kita mungkin bisa mendapatkan metode yang lebih efisien. Ada sejumlah metode yang dapat dibangun yang memiliki sifat ini tetapi berorde lebih tinggi.

### Stabilitas-L

Ternyata tidak semua metode stabil-A sama. Pertimbangkan metode Euler mundur dan metode trapesium. Polinomial stabilitas untuk metode trapesium adalah
$$\begin{aligned}
0 &= U_{n+1} - U_n - \Delta t \frac{1}{2} (\lambda U_n + \lambda U_{n+1}) \\
&= U_{n+1}\left(1 - \frac{1}{2} \Delta t \lambda \right ) - U_n \left(1 + \frac{1}{2}\Delta t \lambda \right) \\
&= \left(\xi - \frac{1 + \frac{1}{2}z}{1 - \frac{1}{2} z}\right) \left(1 - \frac{1}{2} z \right )\\
\end{aligned}$$
yang menunjukkan bahwa itu adalah A-stabil. Mari kita terapkan kedua metode ini pada masalah yang pernah kita lihat sebelumnya dan lihat apa yang terjadi.

```python
import numpy as np                  # Library numerik
import matplotlib.pyplot as plt     # Library plotting

# =========================================
# DEFINISI MASALAH ODE
# =========================================

# Fungsi ODE: du/dt = λ(u - cos(t)) - sin(t)
f = lambda t, lam, u: lam * (u - np.cos(t)) - np.sin(t)

# Solusi eksak (analitik)
u_exact = lambda t_0, eta, lam, t: np.exp(lam * (t - t_0)) * (eta - np.cos(t_0)) + np.cos(t)

# =========================================
# PARAMETER
# =========================================

t_0 = 0.0        # waktu awal
eta = 1.5        # kondisi awal u(t0)
lam = -1e6       # λ sangat besar → stiff problem 😈

t_f = 2.0        # waktu akhir

num_steps = [10, 20, 40, 50]   # jumlah grid

# Array penyimpanan
delta_t = np.empty(len(num_steps))      # Δt
error_euler = np.empty(len(num_steps))  # error Backward Euler
error_trap = np.empty(len(num_steps))   # error trapezoidal

# =========================================
# LOOP PERHITUNGAN
# =========================================

for (i, N) in enumerate(num_steps):
    
    t = np.linspace(0, t_f, N)          # grid waktu
    delta_t[i] = t[1] - t[0]            # langkah waktu
    
    u = u_exact(t_0, eta, lam, t_f)     # solusi eksak di waktu akhir

    # =====================================
    # BACKWARD EULER (implicit Euler)
    # =====================================
    
    U_euler = np.empty(t.shape)
    U_euler[0] = eta                   # kondisi awal
    
    for (n, t_n) in enumerate(t[1:]):
        # Rumus Euler implisit (sudah disusun eksplisit)
        U_euler[n+1] = (U_euler[n]
                        - lam * delta_t[i] * np.cos(t_n)
                        - delta_t[i] * np.sin(t_n)) / (1.0 - lam * delta_t[i])
    
    # Error relatif
    error_euler[i] = np.abs(U_euler[-1] - u) / np.abs(u)

    # =====================================
    # METODE TRAPEZOIDAL (orde 2)
    # =====================================
    
    U_trap = np.empty(t.shape)
    U_trap[0] = eta
    
    for (n, t_n) in enumerate(t[1:]):
        # Metode trapezoid (semi implicit)
        U_trap[n+1] = (U_trap[n]
                        + delta_t[i] * 0.5 * f(t_n, lam, U_trap[n])
                        - 0.5 * lam * delta_t[i] * np.cos(t_n)
                        - 0.5 * delta_t[i] * np.sin(t_n)) / (1.0 - 0.5 * lam * delta_t[i])
    
    # Error absolut
    error_trap[i] = np.abs(U_trap[-1] - u)

# =========================================
# PLOT BACKWARD EULER
# =========================================

fig = plt.figure()
fig.set_figwidth(fig.get_figwidth() * 2)

axes = fig.add_subplot(1, 2, 1)

axes.plot(t, U_euler, 'ro-')                 # solusi numerik
axes.plot(t, u_exact(t_0, eta, lam, t),'k')  # solusi eksak

# =========================================
# ERROR VS Δt (Backward Euler)
# =========================================

axes = fig.add_subplot(1, 2, 2)

axes.loglog(delta_t, error_euler, 'bo')   # error numerik

# fungsi garis referensi orde
order_C = lambda dx, err, order: np.exp(np.log(err) - order * np.log(dx))

# garis orde 1 dan 2
axes.loglog(delta_t,
            order_C(delta_t[1], error_euler[1], 1.0) * delta_t,
            'r--', label="1st Order")

axes.loglog(delta_t,
            order_C(delta_t[1], error_euler[1], 2.0) * delta_t**2,
            'b--', label="2nd Order")

axes.legend(loc=4)
axes.set_title("Comparison of Errors for Backward Euler")
axes.set_xlabel("$\\Delta t$")
axes.set_ylabel("$|U(t_f) - u(t_f)|$")

# =========================================
# PLOT TRAPEZOIDAL
# =========================================

fig = plt.figure()
fig.set_figwidth(fig.get_figwidth() * 2)

axes = fig.add_subplot(1, 2, 1)

axes.plot(t, U_trap, 'ro-')                  # solusi numerik
axes.plot(t, u_exact(t_0, eta, lam, t),'k')  # solusi eksak

# =========================================
# ERROR VS Δt (Trapezoidal)
# =========================================

axes = fig.add_subplot(1, 2, 2)

axes.loglog(delta_t, error_trap, 'bo', label='Trapezoidal')

axes.loglog(delta_t,
            order_C(delta_t[1], error_trap[1], 1.0) * delta_t,
            'r--', label="1st Order")

axes.loglog(delta_t,
            order_C(delta_t[1], error_trap[1], 2.0) * delta_t**2,
            'b--', label="2nd Order")

axes.legend(loc=4)
axes.set_title("Comparison of Errors for Trapezoidal Rule")
axes.set_xlabel("$\\Delta t$")
axes.set_ylabel("$|U(t_f) - u(t_f)|$")

plt.show()
```

Ternyata jika kita melihat metode satu langkah dan mendefinisikan rasio berikut

$$U_{n+1} = R(z) U_n$$

kita dapat mendefinisikan bentuk stabilitas lain, yang disebut **stabilitas-L**, di mana kita mensyaratkan bahwa metode tersebut stabil-A dan bahwa

$$\lim_{z \rightarrow \infty} |R(z)| = 0.$$

Euler mundur stabil-L sedangkan metode trapesium tidak.

### Backward Differencing Formulas

Kelas metode LMM yang berguna untuk masalah ODE kaku adalah metode rumus perbedaan mundur (BDF) yang memiliki bentuk

$$\alpha_0 U_n + \alpha_1 U_{n+1} + \cdots + \alpha_r U_{n+r} = \Delta \beta_r f(U_{n+r})$$

Metode-metode ini dapat diturunkan secara langsung dari perbedaan-perbedaan terhingga yang mundur dari titik $U_{n+r}$ dan titik-titik lainnya di masa lalu. Dengan cara ini, kita dapat memperoleh metode-metode r-langkah yang akurat orde ke-r. Beberapa metode tersebut adalah:

$$\begin{aligned}
    r = 1:& & U_{n+1} - U_n = \Delta t f(U_{n+1}) \\
    r = 2:& &3 U_{n+2} - 4 U_{n+1} + U_n = 2 \Delta t f(U_{n+1}) \\
    r = 3:& &11U_{n+3} - 18U_{n+2} + 9U_{n+1} - 2 U_n = 6 \Delta t f(U_{n+3}) \\
    r = 4:& &25 U_{n+4} - 48 U_{n+3} +36 U_{n+2} -16 U_{n+1} +3 U_n = 12 \Delta t f(U_{n+4})
\end{aligned}$$

### Plotting Wilayah Stabilitas

Jika kita menganggap akar polinomial stabilitas $\xi_j$ sebagai bilangan kompleks dan menuliskannya dalam bentuk eksponensial
$$
\xi_j = |\xi_j| e^{i \theta}.
$$
Di sini $|\xi_j|$ adalah modulus (atau besaran) atau bilangan kompleks dan didefinisikan sebagai $|\xi_j| = x^2 + y^2$ di mana $\xi_j = x + i j$. Jika $\xi_j$ berada di batas wilayah stabilitas absolut maka kita tahu bahwa $|\xi_j| = 1$. Menggunakan ini bersama dengan polinomial stabilitas akan menghasilkan
$$
\rho(e^{i\theta}) - z \sigma(e^{i\theta}) = 0
$$
yang mana penyelesaian untuk $z$ akan menghasilkan
$$
z(\theta) = \frac{\rho(e^{i\theta})}{\sigma(e^{i\theta})}.
$$

Sebagai contoh pertimbangkan metode Adams-Bashforth 2-tahap. Polinomial stabilitas dapat ditemukan sebagai
$$\begin{aligned}
U_{n+2} &= U_{n+1} + \frac{\Delta t}{2} (-f(U_n) + 3 f(U_{n+1})) \\
\pi(\xi, z) &= U_{n+2} - U_{n+1} - \frac{\Delta t}{2} (-f(U_n) + 3 f(U_{n+1})) = 0 \\
&= U_{n+2} - U_{n+1} - \frac{1}{2} (\Delta t \lambda U_n - 3 \Delta t \lambda U_{n+1}) \\
&= 2 \xi^2 - 2 \xi + 3 z\xi - z \\
&= \rho(\xi, z) + z \sigma(\xi, z)
\end{aligned}$$
dimana
$$
\rho(\xi, z) = 2 ( \xi - 1) \xi ~~~ \text{dan} ~~~ \sigma(\xi, z) = 3 \xi - 1
$$
sehingga
$$
z(\theta) = \frac{2 (\xi - 1) \xi}{3 \xi - 1}.
$$

Hal ini tidak serta merta memastikan bahwa jika diberikan $\theta$ maka $z(\theta)$ akan terletak pada batas wilayah stabilitas absolut. Hal ini dapat terjadi ketika $\xi_j = 1$ tetapi di sebelah kiri dan kanan kurva $\xi_j > 1$ dan karenanya tidak menandai batas wilayah tersebut. Untuk menentukan apakah wilayah tertentu yang digariskan oleh kurva ini berada di dalam atau di luar wilayah stabilitas, kita dapat mengevaluasi semua akar $\pi(\xi, z)$ pada beberapa $z$ di dalam wilayah yang dimaksud dan melihat apakah $\forall j, \xi_j < 1$.

Untuk metode satu langkah, hal ini menjadi lebih mudah jika kita melihat rasio $R(z)$ yang kita definisikan sebelumnya sebagai
$$
U_{n+1} = R(z) U_n
$$
dalam kasus metode deret Taylor orde p yang diterapkan pada $u'(t) = \lambda u$, kita peroleh

$$\begin{aligned}
U_{n+1} &= U_n + \Delta t \lambda U_n + \frac{1}{2}\Delta t^2 \lambda^2 U_n + \cdots + \frac{1}{p!}\Delta t^p \lambda^p U_n \\
&=\left(1 + z + \frac{1}{2} z^2 + \frac{1}{6} z^3 + \cdots +\frac{1}{p!}z^p\right) U_n \Rightarrow\\
R(z) &= 1 + z + \frac{1}{2} z^2 + \frac{1}{6} z^3 + \cdots +\frac{1}{p!}z^p.
\end{aligned}$$ 

Menetapkan $R(z) = e^{i\theta}$ dapat mengarah pada cara untuk memecahkan batas tetapi (di mana $|R(z)| = 1$) tetapi ini sangat sulit dilakukan secara umum. Sebaliknya jika kita memplot kontur $R(z)$ di dataran kompleks, kita dapat memilih kontur $R(z)=1$ dan memplotnya.

```python
import numpy as np                  # Library numerik (array, bilangan kompleks, dll)
import matplotlib.pyplot as plt     # Library untuk plotting

# =========================================
# PARAMETER UMUM
# =========================================

theta = np.linspace(0.0, 2.0 * np.pi, 100)  
# Sudut untuk membentuk lingkaran (dipakai nanti di metode lain)

# =========================================
# FORWARD EULER
# =========================================

fig = plt.figure()                                  # Membuat figure
fig.set_figwidth(fig.get_figwidth() * 2.0)           # Memperlebar figure
fig.set_figheight(fig.get_figheight() * 2.0)         # Mempertinggi figure

axes = fig.add_subplot(2, 2, 1)                     # Subplot posisi 1

# Grid di bidang kompleks (x = real, y = imajiner)
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y                                       # Bilangan kompleks Z = x + iy

# Uji stabilitas (contoh titik)
print("Forward Euler's method - Stability region tests")
z = -1 + 1j * 0
print("  Inside of circle: ", np.abs(1.0 + z))       # harus ≤ 1 untuk stabil
z = -3 + 1j * 0
print("  Outside of circle: ", np.abs(1.0 + z))

# Plot batas stabilitas: |1 + z| = 1
axes.contour(X, Y, np.abs(1.0 + Z), levels=[1.0])

# Plot sumbu
axes.plot(x, np.zeros(x.shape),'k')                  # sumbu x
axes.plot(np.zeros(y.shape), y,'k')                  # sumbu y

axes.set_aspect('equal')                             # skala sama
axes.set_title("Forward Euler")

# =========================================
# BACKWARD EULER
# =========================================

axes = fig.add_subplot(2, 2, 2)

x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Uji stabilitas
print("Backward Euler's method - Stability region tests")
z = 1 + 1j * 0
print("  Inside of circle: ", np.abs(1.0 + z))
z = -1 + 1j * 0
print("  Outside of circle: ", np.abs(1.0 + z))

# Plot batas stabilitas: |1 / (1 - z)| = 1
axes.contour(X, Y, np.abs(1.0 / (1.0 - Z)), levels=[1.0])

axes.plot(x, np.zeros(x.shape),'k')
axes.plot(np.zeros(y.shape), y,'k')

axes.set_aspect('equal')
axes.set_title("Backwards Euler")

# =========================================
# METODE TAYLOR ORDE 4
# =========================================

axes = fig.add_subplot(2, 2, 3)

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Uji stabilitas
print("Taylor series method of order 4 - Stability region tests")
z = -1 + 1j * 0
print("  Inside of strange region: ", np.abs(1.0 + z))
z = -4 + 1j * 0
print("  Outside of strange region: ", np.abs(1.0 + z))

# Fungsi stabilitas Taylor orde 4:
# R(z) = 1 + z + z^2/2 + z^3/6 + z^4/24
axes.contour(X, Y,
             np.abs(1 + Z + 0.5 * Z**2 + (1/6) * Z**3 + (1/24) * Z**4),
             levels=[1.0])

axes.plot(x, np.zeros(x.shape),'k')
axes.plot(np.zeros(y.shape), y,'k')

axes.set_aspect('equal')
axes.set_title("4th Order Taylor Series")

# =========================================
# 2-STEP ADAMS-BASHFORTH
# =========================================

theta = np.linspace(0.0, 2.0 * np.pi, 1000)
xi = np.exp(1j * theta)   # Titik di lingkaran satuan (|ξ| = 1)

# Fungsi karakteristik metode
rho_2AB = lambda xi: (xi - 1.0) * xi
sigma_2AB = lambda xi: (3.0 * xi - 1.0) / 2.0

# Transformasi ke bidang z
z_2AB = rho_2AB(xi) / sigma_2AB(xi)

axes = fig.add_subplot(2, 2, 4)

axes.plot(z_2AB.real, z_2AB.imag)    # Plot batas stabilitas

axes.plot(x, np.zeros(x.shape),'k')
axes.plot(np.zeros(y.shape), y,'k')

axes.set_title("2-step Adams-Bashforth")
axes.set_aspect('equal')

axes.set_xlim([-2, 3])
axes.set_ylim([-2, 2])

# =========================================
# TAMPILKAN SEMUA PLOT
# =========================================

plt.show()
```
